# 03_02_Regional_Boundary

## 1. Introduction

The regional boundary defines the spatial extent of the study area.

It is used to:
- Clip all spatial datasets to a consistent area
- Provide a reference frame for analysis
- Support reuse of the workflow for other regions by replacing the dataset

**Data Source: S1**

To adapt this notebook to a different location, simply replace the boundary GeoJSON file in the `data/` directory and update the file path in the code below.

## 2. Imports

Import core geospatial and visualization libraries used in this notebook:


```python
import geopandas as gpd 
import matplotlib.pyplot as plt
```

### 3. Loading the Regional Boundary

```python
boundary_file = "../../data/Regional_Boundary.geojson"

boundary = gpd.read_file(boundary_file)

boundary.head()
```

### 4. Boundary Coordinate Information

Before displaying the study area it is useful to examine the coordinate system and spatial extent of the regional boundary.

These values define the geographic extent of the study area and will later be used when clipping elevation, hydrology, and environmental datasets.

Different regions may use different coordinate systems, so these values should always be verified before combining datasets.

```python
print("Coordinate System:")
print(boundary.crs)

print()

print("Bounds:")
print(boundary.total_bounds)
```

```python
minx, miny, maxx, maxy = boundary.total_bounds

print("Minimum Longitude:", minx)
print("Minimum Latitude:", miny)

print()

print("Maximum Longitude:", maxx)
print("Maximum Latitude:", maxy)
```

### 5. Regional Boundary Output

```python
fig, ax = plt.subplots(figsize=(8,8))

boundary.plot(
    ax=ax,
    edgecolor="black",
    facecolor="lightblue",
    alpha=0.5
)

ax.set_title("Waterloo Region Boundary")

ax.set_axis_off()

plt.show()
```
