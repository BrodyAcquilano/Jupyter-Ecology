### S2 — Elevation Dataset (CDEM Derived Products)
Canadian Digital Elevation Model (CDEM) extraction including:
- DEM (elevation raster)
- Hillshade
- Slope
- Aspect
- Color relief
- Point elevation data

Used for terrain modelling and hydrological surface analysis.

Source: Canadian GeoHub:

https://app.geo.ca/en-ca/map-browser/record/7f245e4d-76c2-4caa-951a-45d1d2051333

The government uses a geospatial data extraction service. Users are directed to this service to extract data so that users don't have to download the full dataset which would be too large and take up a lot of storage space and take a long time to download. You can make a request at the Government of Canada's Geospatial Data Extraction Service: 

https://maps.canada.ca/czs/index-en.html

I have made a request and extracted data for an area slightly larger than Waterloo Region which I will clip to the regional boundary. 

Files:
- DEM.tif
- HILLSHADE.tif
- COLOR_HILLSHADE.tif
- SLOPE.tif
- ASPECT.tif
- COLOR.tif
- POINT_DATA.xyz
