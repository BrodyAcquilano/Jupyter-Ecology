# Water Flow

## 1. Introduction

Water movement across the landscape is primarily governed by terrain elevation and slope.

This notebook builds on the elevation surface developed in the set of notebooks on elevation and introduces a basic hydrological interpretation of that surface. Instead of focusing on static terrain structure, the goal here is to begin examining how water would conceptually move across the landscape based on changes in elevation.

The objectives of this section are:

- Compute terrain gradients from elevation data.
- Estimate relative flow potential across the surface.
- Visualize likely drainage patterns.
- Explore directional flow behavior.
- Identify areas of potential accumulation or convergence.

These methods provide a simplified representation of surface hydrology. They do not simulate full fluid dynamics, but instead approximate how water would respond to terrain under gravity-driven surface runoff assumptions.

This forms the foundation for later notebooks that will extend into watershed delineation, flow accumulation modelling, and flood-risk interpretation.
