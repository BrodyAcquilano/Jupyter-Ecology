# 01. Methodologies for Analyzing Plant Illness

## From Initial Observation to Evidence, Diagnosis,
## Population Analysis, and Treatment

Plant illness is rarely explained by one visible symptom.

A damaged leaf, broken branch, patch of missing bark,
fungal growth, insect hole, or declining crown may have
many possible causes.

Some symptoms indicate serious threats to the survival or
structural stability of a tree. Other symptoms may be
temporary, cosmetic, seasonal, or part of ordinary
ecological interactions.

This notebook develops a general methodology for
investigating plant illness.

The central question is:

$$
\text{How do we move from an initial observation}
$$

$$
\text{to a defensible conclusion?}
$$

The investigation will consider:

- how plant illness should be defined;
- how plants resist infection and recover from injury;
- the major biological and environmental causes of decline;
- how symptoms and signs can be observed systematically;
- how competing explanations can be compared;
- how the seriousness of an issue can be evaluated;
- how individual plants and populations can be sampled;
- how disease incidence and severity can be measured;
- how spatial and temporal patterns can be analyzed;
- how treatments or management responses can be evaluated;
- what conclusions are possible with limited resources.

This introductory notebook does not attempt to diagnose a
particular tree or complete a full field study.

Instead, it establishes the conceptual and methodological
foundation for later notebooks involving field observations,
tree inventories, maps, photographs, statistical analysis,
and possible treatment strategies.

## 1. Purpose of the Study

The purpose of this project is to develop a repeatable method
for studying plant illness in urban environments.

The project may begin with observations such as:

- dark spots on leaves;
- missing leaf tissue;
- discoloured foliage;
- broken or hanging branches;
- dead branches;
- missing bark;
- wounds on trunks or roots;
- exposed or severed roots;
- unusual fungal structures;
- web-like material;
- insect holes;
- declining or sparse crowns;
- new shoots growing from trunks or branches.

These observations may suggest that a plant is unhealthy,
but they do not automatically identify the cause.

The methodology must distinguish between four stages:

$$
\text{Observation}
\rightarrow
\text{Description}
\rightarrow
\text{Analysis}
\rightarrow
\text{Conclusion}
$$

An **observation** records what is visible.

A **description** converts the observation into measurable
characteristics.

An **analysis** compares possible explanations using
additional evidence.

A **conclusion** states what the available evidence supports,
including any remaining uncertainty.

The goal is not merely to name diseases.

The goal is to understand:

- what happened;
- how serious it is;
- how common it is;
- what evidence supports the explanation;
- what additional evidence is needed;
- what actions may be appropriate.

## 2. Observation Is Not Diagnosis

A visible symptom is evidence that something has occurred.

It is not necessarily evidence of one specific cause.

For example, a hanging branch might have resulted from:

- wind;
- ice;
- mechanical impact;
- poor branch attachment;
- internal decay;
- pruning;
- vandalism;
- animal activity;
- a previous infection;
- several interacting causes.

A hole in a leaf might have resulted from:

- insect feeding;
- physical tearing;
- hail;
- abrasion;
- dead tissue falling away;
- a bacterial lesion;
- a fungal lesion;
- ordinary accumulated damage.

A responsible investigation must avoid moving directly from:

$$
\text{I observed damage}
$$

to:

$$
\text{I know what caused the damage}
$$

Instead, the investigation should follow:

$$
\boxed{
\text{Observed condition}
\rightarrow
\text{possible explanations}
\rightarrow
\text{additional evidence}
\rightarrow
\text{provisional conclusion}
}
$$

The word **provisional** is important.

A field diagnosis may remain uncertain unless it is supported
by expert inspection, microscopy, culturing, DNA analysis,
soil testing, chemical analysis, or another specialized method.

## 3. Signs, Symptoms, Injuries, and Causes

Several terms must be kept separate.

### Symptom

A symptom is a response of the plant.

Examples include:

- yellowing;
- wilting;
- leaf spots;
- dead tissue;
- early leaf loss;
- crown thinning;
- branch dieback;
- cracking;
- swelling;
- reduced growth.

### Sign

A sign is direct physical evidence of a biological agent.

Examples may include:

- fungal fruiting bodies;
- spores;
- visible mycelium;
- bacterial ooze;
- insect larvae;
- insect eggs;
- feeding galleries;
- shed insect skins;
- frass.

### Injury

An injury is physical or chemical damage that may not involve
a living pathogen.

Examples include:

- broken branches;
- cuts;
- girdling;
- salt injury;
- heat injury;
- frost injury;
- root severing;
- chemical exposure;
- vehicle impact.

### Cause

A cause is the process or agent responsible for the condition.

### Diagnosis

A diagnosis is a reasoned identification of the cause based on
the available evidence.

A diagnosis should be consistent with:

- the host species;
- the affected tissue;
- the symptom pattern;
- the season;
- the location;
- the environmental conditions;
- the likely transmission pathway;
- the progression of the condition through time.

## 4. What Does Plant Illness Mean?

A plant may be considered unhealthy when its ability to
survive, grow, reproduce, maintain its structure, or perform
normal physiological functions has been reduced.

However, visible damage does not always produce a meaningful
reduction in overall health.

Plants naturally experience cycles of:

- growth;
- dormancy;
- leaf production;
- leaf loss;
- flowering;
- fruiting;
- branch development;
- tissue death;
- tissue replacement.

A deciduous tree losing its leaves in autumn is not ill.

A tree may also lose leaves, twigs, or small branches during
stress and still survive.

Visible tissue loss must therefore be interpreted according
to:

- the plant species;
- the season;
- the tissue affected;
- the amount of tissue affected;
- the rate of change;
- the plant's ability to replace the tissue.

For example:

- a tree may lose some leaves and remain healthy;
- a tree may experience environmental stress and still bounce back;
- a leaf disease may appear dramatic but remain cosmetic;
- a tree may replace damaged foliage;
- one branch may die while the rest of the tree survives;
- a root disease may be serious before leaf symptoms appear;
- internal decay may create risk without immediately killing
  the tree;
- seasonal dormancy may resemble decline to an inexperienced
  observer.

Plants are often highly resilient.

Unlike animals with a largely fixed adult body plan, woody
plants continue producing new leaves, roots, shoots, branches,
and protective tissues throughout their lives.

This continuing growth allows a tree to:

- replace lost leaves;
- activate dormant buds;
- redirect growth;
- abandon damaged tissue;
- isolate some injuries;
- survive partial loss of its crown or roots.

The survival of a plant species also depends on more than the
survival of one individual.

Plants persist through:

- seed production;
- vegetative reproduction;
- dispersal;
- long life spans;
- repeated seasonal growth;
- large populations;
- genetic variation.

However, resilience is not unlimited.

A plant may approach a threshold beyond which damage,
infection, or resource loss can no longer be compensated for.

Conceptually:

$$
\text{Plant Condition}
=
\text{Damage}
-
\text{Defence and Recovery Capacity}
$$

A serious decline becomes more likely when:

$$
\text{Damage Rate}
>
\text{Recovery Rate}
$$

This threshold will differ among species and tissues.

Repeated loss of deciduous leaves may be survivable because
the tree can produce a new canopy during a later season.

Many conifers retain needles for several years and may replace
lost foliage more slowly. Severe needle loss may therefore
have different consequences.

Similarly, loss of a few surface roots is not equivalent to
extensive root decay, and loss of bark on one side of a trunk
is not equivalent to damage extending around most of its
circumference.

Plant health should therefore be treated as a
multidimensional condition.

A conceptual model is:

$$
H_i
=
f(
L_i,
C_i,
B_i,
T_i,
R_i,
G_i,
P_i,
E_i
)
$$

Where:

- $H_i$ = overall condition of plant $i$;
- $L_i$ = leaf condition;
- $C_i$ = crown condition;
- $B_i$ = branch condition;
- $T_i$ = trunk or stem condition;
- $R_i$ = root condition;
- $G_i$ = growth and regeneration;
- $P_i$ = pests or pathogens;
- $E_i$ = environmental conditions.

This is a conceptual relationship rather than a validated
health equation.

It shows that whole-plant health cannot be determined from one
damaged leaf or one visible organism.

## 5. The Plant Disease Triangle

A traditional model of plant disease considers three
interacting components:

1. a susceptible host;
2. a harmful agent;
3. an environment that supports disease.

This relationship can be written as:

$$
\text{Disease}
=
f(
\text{Host},
\text{Agent},
\text{Environment}
)
$$

For an urban tree study, the model can be expanded:

$$
D_i(t)
=
f(
H_i,
A_i,
E_i(t),
W_i(t),
V_i(t),
M_i(t)
)
$$

Where:

- $D_i(t)$ = disease or damage outcome;
- $H_i$ = host characteristics;
- $A_i$ = biological or non-biological agent;
- $E_i(t)$ = environmental conditions;
- $W_i(t)$ = wounds and structural damage;
- $V_i(t)$ = vectors and transmission pathways;
- $M_i(t)$ = maintenance and management history.

This model recognizes that an urban tree may be influenced by:

- planting conditions;
- road salt;
- pruning;
- restricted roots;
- construction;
- irrigation;
- drought;
- insects;
- fungi;
- bacteria;
- human activity.

Several factors may act at the same time.

## 6. Plant Barriers and Defence Mechanisms

Plants are not passive surfaces on which organisms grow.

They possess physical, chemical, physiological, and ecological
barriers that may prevent infection or limit damage.

### Physical Barriers

Physical barriers include:

- bark;
- waxy leaf cuticles;
- epidermal tissue;
- cell walls;
- protective bud scales;
- sealed tissue surfaces.

These structures make it more difficult for pathogens to
reach vulnerable internal tissues.

### Chemical Defences

Plants may produce compounds that:

- inhibit fungal growth;
- discourage insect feeding;
- damage invading organisms;
- signal nearby cells;
- strengthen cell walls;
- isolate infected tissue.

### Localized Defence

Plants may isolate or sacrifice a small area to limit spread.

Possible responses include:

- localized cell death;
- formation of barriers around damaged wood;
- wound sealing;
- compartmentalization of decay;
- shedding infected leaves;
- abandoning damaged twigs or branches.

### Regrowth and Compensation

Plants may compensate for tissue loss through:

- new leaf production;
- dormant bud activation;
- new shoot growth;
- increased growth in surviving tissue;
- redistribution of stored resources;
- replacement of damaged roots or branches.

New growth does not automatically prove that the plant is
responding to disease.

It may also be ordinary seasonal growth.

### Wound Closure and Compartmentalization

Trees do not heal wounds in exactly the same way that animals
repair injured tissue.

Damaged wood generally remains damaged.

Instead, a tree may:

- form barriers around the injured area;
- limit the movement of decay;
- produce new wood around the wound;
- grow callus tissue inward from the wound edges;
- cover the injury gradually;
- redirect transport through surviving tissue.

This process is often described as compartmentalization.

A wound may eventually become covered by new growth without
the original damaged wood being restored.

Whether closure succeeds depends on:

- wound size;
- wound depth;
- tree species;
- tree age;
- tree vigour;
- moisture;
- the location of the injury;
- continued infection or decay;
- the amount of living tissue around the wound.

Small wounds may be closed successfully.

Large wounds, repeated injuries, trunk cavities, and wounds
containing active decay may remain open or structurally weak.

A tree that has grown over a wound is therefore not
necessarily identical to an undamaged tree.

## 7. Fungi, Bacteria, and Plant Microbial Communities

Plants and fungi have interacted and co-evolved over extremely
long periods of evolutionary time.

Fungi are natural components of:

- soil;
- leaf litter;
- roots;
- bark;
- wood;
- living plant tissue;
- dead plant tissue.

Many plants form beneficial associations with mycorrhizal
fungi.

These fungi may assist plants with:

- water uptake;
- phosphorus and nutrient uptake;
- tolerance of environmental stress;
- interactions with soil microbial communities.

Other fungi live harmlessly on or inside plants.

Some fungi decompose dead organic material.

A smaller group acts as plant pathogens.

The presence of fungi should therefore not automatically cause
alarm.

A fungus associated with a tree may be:

- beneficial;
- neutral;
- an endophyte;
- a decomposer;
- an opportunist;
- a secondary pathogen;
- a primary pathogen.

Fungi nevertheless cause a large proportion of known plant
diseases.

This does not mean that fungi are simply designed to attack
plants.

Fungi are especially successful plant-associated organisms
because many species can:

- produce large numbers of spores;
- disperse through air, water, soil, or organisms;
- grow as branching hyphae through suitable material;
- use enzymes that break down plant structural compounds;
- colonize living, dying, or dead organic material;
- persist in soil or plant debris;
- specialize on particular plant hosts.

Different pathogenic fungi use different feeding strategies.

### Biotrophic Fungi

Biotrophic fungi obtain resources from living host tissue and
may manipulate the host while keeping much of it alive.

### Necrotrophic Fungi

Necrotrophic fungi kill host tissue and obtain resources from
the dead material.

### Hemibiotrophic Fungi

Hemibiotrophic fungi begin by colonizing living tissue and
later kill part of the host.

Bacteria also cause important plant diseases.

Examples of bacterial effects include:

- leaf spots;
- blights;
- wilts;
- soft rots;
- cankers;
- galls;
- vascular obstruction.

Bacterial diseases are not unimportant simply because fewer
may appear in a forest-disease database.

The apparent dominance of fungi may reflect:

- the genuine diversity of fungal plant pathogens;
- the ability of fungi to digest structural plant material;
- the visibility of fungal fruiting bodies;
- the importance of fungal diseases in forests;
- the purpose and scope of a particular database;
- differences in how fungi, bacteria, and viruses are detected.

Bacteria are microscopic and may colonize internal tissue
without producing a visible structure comparable to a
mushroom or fungal fruiting body.

Laboratory testing may therefore be needed to distinguish a
bacterial disease from fungal, viral, or environmental damage.

Plants also interact with beneficial bacteria.

Some bacteria may:

- compete with pathogens;
- produce antimicrobial compounds;
- assist nutrient cycling;
- occupy possible infection sites;
- stimulate plant defence;
- support root growth.

Plant health is therefore partly influenced by an ecological
community rather than only by a conflict between one plant
and one pathogen.

## 8. How Plant Barriers Are Bypassed

Pathogens and pests may reach plant tissue through several
routes.

Possible entry pathways include:

- natural openings in leaves;
- young or developing tissue;
- damaged roots;
- broken branches;
- pruning wounds;
- cracks in bark;
- insect feeding wounds;
- frost cracks;
- storm damage;
- mechanical injuries;
- infected roots;
- contaminated soil;
- contaminated water;
- infected plant material;
- contaminated tools.

The presence of a wound may increase the chance of infection.

However, a wound alone does not prove that infection occurred.

Disease near a wound also does not prove that the wound
happened first.

Several sequences are possible:

$$
\text{Wound}
\rightarrow
\text{infection}
\rightarrow
\text{decline}
$$

$$
\text{Infection}
\rightarrow
\text{weakened tissue}
\rightarrow
\text{branch failure}
$$

$$
\text{Environmental stress}
\rightarrow
\text{reduced defence}
\rightarrow
\text{infection}
\rightarrow
\text{structural damage}
$$

Determining the order of events may require repeated
observations, maintenance records, weather data, or
professional assessment.

### A Wound Does Not Guarantee Infection

A wound creates an opportunity, but infection still depends on:

- whether a pathogen reaches the wound;
- whether the pathogen can infect that host;
- whether environmental conditions support infection;
- whether the tissue remains exposed;
- whether the plant can isolate the affected area;
- whether competing microorganisms suppress the pathogen.

This means that neither of the following assumptions is safe:

$$
\text{Wound}
=
\text{Infection}
$$

$$
\text{Fungus near wound}
=
\text{Cause of wound}
$$

The wound, infection, and visible fungal growth may represent
different stages of one process, or they may be unrelated.

This prevents the study from assuming that every abnormality

### Pathogens

- fungi;
- fungus-like organisms;
- bacteria;
- viruses;
- viroids;
- phytoplasmas;
- nematodes;
- parasitic plants.

### Insects and Other Animals

- leaf-chewing insects;
- sap-feeding insects;
- bark beetles;
- wood-boring insects;
- leaf miners;
- gall-forming organisms;
- mites;
- caterpillars;
- vertebrate browsing;
- bark stripping;
- nesting or burrowing activity.

Some insects directly damage plants.

Others act as vectors that carry pathogens.

- air pollution;

### Physical Injury

Possible sources of physical injury include:

- storms;
- falling objects;
- vehicles;
- bicycles;
- construction equipment;
- lawn-maintenance equipment;
- pruning;
- staking materials;
- wires;
- signs;
- drilling;
- cutting;
- root excavation;
- pedestrian traffic;
- animal activity;
- vandalism.

### Planting and Management Conditions

Management-related factors include:

- poor pruning technique;
- delayed pruning;
- incorrect planting depth;
- restricted planting pits;
- excessive mulch;
- inadequate mulch;
- inadequate watering;
- excessive watering;
- damaged nursery stock;
- poor species selection;
- repeated disturbance.

### Urban Cultivation, Transplanting, and Root Restriction

Urban trees often begin life under conditions that differ from
those of naturally established trees.

Trees grown for transportation and planting may be produced
as:

- container-grown stock;
- bare-root stock;
- balled-and-burlapped stock;
- field-grown nursery stock.

Moving a tree usually reduces or disturbs its root system.

Some root pruning is intentional and may be used to produce a
compact root ball with new fine roots before transportation.

However, transplanting can still leave the tree with a much
smaller absorbing root system than an established tree of the
same size.

Possible consequences include:

- reduced water uptake;
- reduced nutrient uptake;
- transplant shock;
- slower shoot growth;
- crown dieback;
- reduced stored resources;
- greater sensitivity to drought;
- delayed establishment.

A cultivated tree is not automatically genetically weak or
immunologically defective.

The more defensible conclusion is that cultivation,
transplanting, and urban planting conditions may create
physiological stress that reduces the resources available for
defence and recovery.

After planting, root development may also be restricted by:

- compacted soil;
- small planting pits;
- pavement;
- buried utilities;
- poor drainage;
- low soil oxygen;
- circling roots;
- girdling roots;
- repeated excavation;
- road salt;
- limited water;
- incompatible soil layers.

Roots also interact with fungi, bacteria, neighbouring roots,
and chemical signals in the soil.

A restricted or disturbed root zone may alter these
relationships.

It would be too strong to conclude from appearance alone that
an urban tree has lost its ability to communicate.

Instead, the study may investigate whether restricted rooting
space and disturbed soil reduce the tree's access to normal
below-ground biological interactions.

This creates a possible causal sequence:

$$
\text{Root Loss or Restriction}
\rightarrow
\text{Water and Nutrient Stress}
\rightarrow
\text{Reduced Growth and Defence}
\rightarrow
\text{Greater Susceptibility}
$$

## 11. Normal or Low-Concern Processes

Not every unusual feature is evidence of serious illness.

Possible low-concern explanations include:

- ordinary leaf aging;
- seasonal leaf loss;
- minor insect feeding;
- old pruning wounds;
- natural bark shedding;
- branch self-pruning;
- harmless lichen;
- moss;
- minor cosmetic leaf disease;
- decomposition of tissue that was already dead.

This creates an important distinction:

$$
\text{Visible abnormality}
\neq
\text{serious whole-plant illness}
$$

The study must determine whether the condition affects:

- appearance only;
- temporary growth;
- long-term growth;
- reproduction;
- survival;
- structural stability;
- other nearby plants.

## 12. Primary, Secondary, and Opportunistic Agents

The organism most visible on a damaged tree may not be the
original cause of decline.

### Primary Agent

A primary agent can attack relatively healthy tissue and begin
the condition.

### Secondary Agent

A secondary agent attacks tissue that has already been
weakened by another stressor.

### Opportunistic Organism

An opportunistic organism takes advantage of damaged, dying,
or dead tissue.

### Decomposer

A decomposer breaks down tissue after it has died.

A possible sequence is:

$$
\text{Drought}
\rightarrow
\text{reduced defence}
\rightarrow
\text{insect attack}
\rightarrow
\text{branch death}
\rightarrow
\text{fungal decomposition}
$$

In this example, the visible fungus may not have caused the
original decline.

The investigation should ask:

- Did the organism attack healthy tissue?
- Was the tissue already injured?
- Was the plant already stressed?
- Is the organism spreading into living tissue?
- Is the organism restricted to dead material?
```

```
## 13. Transmission Pathways and Vectors

A vector is an organism that carries a pathogen from one host
to another.

Possible vectors include:

- beetles;
- aphids;
- leafhoppers;
- mites;
- nematodes;
- birds;
- mammals;
- humans.

Not every transmission pathway requires a living vector.

Other pathways include:

- wind;
- rain splash;
- surface water;
- connected roots;
- soil movement;
- contaminated tools;
- nursery stock;
- transported firewood;
- fallen leaves;
- infected seeds;
- movement of branches or wood debris.

Field observations should record both:

- **Suspected Agent**
- **Suspected Transmission or Damage Pathway**

This prevents wind, water, or tools from being incorrectly
classified as biological vectors.

### The Central Importance of Roots

Roots are sometimes compared with a digestive system because
they are responsible for much of the plant's acquisition of
water and mineral nutrients.

The comparison is useful, but incomplete.

Roots also:

- anchor the plant;
- store carbohydrates;
- produce and respond to chemical signals;
- interact with fungi and bacteria;
- transport water and dissolved minerals;
- exchange gases with the soil;
- produce new absorbing tissue.

Many major plant-health problems begin in or around the roots.

Examples include:

- root rot;
- waterlogging;
- drought;
- nutrient deficiency;
- nutrient toxicity;
- soil compaction;
- low soil oxygen;
- nematodes;
- root-feeding insects;
- girdling roots;
- severed roots;
- chemical contamination;
- excessive fertilizer;
- planting too deeply;
- buried root collars.

Root problems may later produce symptoms elsewhere, including:

- wilting;
- small leaves;
- yellowing;
- reduced growth;
- crown thinning;
- branch dieback;
- instability;
- increased vulnerability to other agents.

However, it would be inaccurate to state that most visible
plant problems must begin in the roots.

Plants may also be attacked directly through:

- leaves;
- flowers;
- fruit;
- buds;
- shoots;
- branches;
- bark;
- wounds;
- vascular tissue.

Roots should be treated as a major diagnostic category, not
as the automatic explanation for every symptom.

A useful diagnostic question is:

$$
\text{Could the visible symptom be a direct above-ground response}
$$

$$
\text{to an unseen below-ground problem?}
$$

or

$$
\text{Can it be traced back to the roots through a chain of events?}
$$

## 15. Diagnostic Question Tree

The diagnostic process can be approached like a dichotomous
key.

It does not immediately identify one disease.

Instead, it narrows the possible causal categories.

### Question 1: Is the Feature Actually Abnormal?

Ask:

- Is it present on healthy plants of the same species?
- Is it normal for the season?
- Is it ordinary bark texture or leaf shape?
- Is it old damage rather than active damage?
- Is it lichen, moss, or another harmless organism?

### Question 2: Which Tissue Is Affected?

Identify the primary location:

- leaves;
- buds;
- shoots;
- branches;
- trunk;
- bark;
- roots;
- flowers;
- fruit;
- the entire plant.

### Question 3: Is the Damage Localized or Widespread?

Determine whether the condition affects:

- one leaf;
- one branch;
- one side of the crown;
- the entire crown;
- one tree;
- several nearby trees;
- several species;
- an entire site.

- connected root systems.

### Question 8: Is the Condition Progressing?

Repeated observations should determine whether it is:

- stable;
- spreading;
- recovering;
- recurring seasonally;
- appearing on additional trees;
- producing new structural problems.

### Seasonal Timing and Plant Phenology

The timing of an observation can be as important as its
appearance.

Plants pass through seasonal developmental stages known as
phenological stages.

These may include:

- bud swelling;
- bud break;
- leaf emergence;
- flowering;
- fruit production;
- shoot extension;
- leaf colour change;
- leaf fall;
- dormancy.

A structure or symptom may be normal during one stage and
unusual during another.

Pathogens and insects also have seasonal life cycles.

A diagnostic record should therefore include:

- observation date;
- host developmental stage;
- recent weather;
- when the symptom first appeared;
- whether the symptom is new or left from a previous season;
- whether the agent is expected to be active at that time.

Seasonal observations can help distinguish between:

- ordinary leaf emergence and replacement growth;
- natural autumn leaf loss and premature defoliation;
- frost injury and infectious blight;
- old dead tissue and an expanding current infection;
- spring fungal activity and late-season symptoms.

The visible symptom may also appear long after the original
infection occurred.

For example, a plant may be infected during wet spring
conditions but develop obvious lesions later in the growing
season.

This means that:

$$
\text{Date Symptoms Are Observed}
\neq
\text{Date Infection Began}
$$

Repeated observations across the growing season are therefore
more informative than one visit.

## 17. Building Competing Hypotheses

Every important observation should generate several plausible
explanations.

For a partially detached branch, hypotheses may include:

1. wind damage;
2. ice damage;
3. mechanical impact;
4. a weak branch attachment;
5. internal decay;
6. incomplete pruning;
7. improper pruning;
8. animal damage;
9. intentional damage;
10. several causes acting together.

The purpose is not to list possibilities forever.

The purpose is to identify evidence that separates them.

A useful hypothesis record may contain:

| Field | Description |
|---|---|
| Hypothesis | Proposed explanation |
| Supporting Evidence | Evidence consistent with it |
| Contradicting Evidence | Evidence against it |
| Missing Evidence | Information still required |
| Confidence | Low, moderate, or high |

Unusual explanations may remain possible.

However, they should not receive more weight than the
available evidence supports.

### Accessible Diagnosis and Reference Tools

Not every useful plant-health investigation requires advanced
laboratory equipment.

Careful qualitative observations can narrow the possibilities
substantially when they are combined with existing research.

Useful low-cost tools include:

- a notebook or digital field form;
- a camera;
- a ruler or scale;
- a hand lens;
- a tree-identification guide;
- a disease glossary;
- a symptom index;
- maps;
- weather records;
- municipal tree inventories;
- repeated visits;
- trusted regional reference databases.

The Government of Canada's database:

[Trees, insects, mites, and diseases of Canada's forests](
https://tidcf.nrcan.gc.ca/en
)

contains information on:

- tree hosts;
- insects;
- mites;
- diseases caused by pathogens;
- geographic distributions;
- signs and symptoms;
- biological information;
- photographs.

The alphabetical disease index is available at:

[Diseases caused by pathogens](
https://tidcf.nrcan.gc.ca/en/diseases-caused-by-pathogens?index=a#index
)

A reference database can help generate or narrow hypotheses.

It cannot replace confirmation of:

- host species;
- geographic range;
- affected tissue;
- seasonal timing;
- characteristic signs;
- similar disorders.

A useful field diagnosis may be possible when a condition has
a distinctive combination of host, timing, distribution, and
symptoms.

However, visually similar diseases may be caused by different
species, strains, or even different categories of agents.

The conclusion should match the available level of evidence.

For example:

- **leaf-spot-like symptom**;
- **consistent with tar spot**;
- **probable tar spot**;
- **confirmed tar spot caused by a named species**.

These statements represent different levels of confidence.

## 18. Determining Whether an Issue Is Serious

The visual size of a symptom is not always equal to its
biological seriousness.

Seriousness should be evaluated using several dimensions.

### Tissue Importance

Damage to different tissues has different consequences.

Leaf damage may reduce photosynthesis but may be temporary.

Root, vascular, trunk, and structural damage may have more
serious long-term consequences.

### Extent

Possible measurements include:

- percentage of leaf area damaged;
- percentage of leaves affected;
- percentage of crown affected;
- percentage of branches dead;
- percentage of trunk circumference damaged;
- number of exposed or severed roots.

### Rate of Progression

Rapidly expanding damage may require more concern than stable,
old damage.

### Structural Risk

A tree may remain alive while becoming structurally unsafe.

Potential concerns include:

- large hanging branches;
- major cavities;
- extensive decay;
- root instability;
- trunk cracks;
- weak branch unions;
- substantial bark loss.
```

```
## 19. A Provisional Seriousness Scale

A preliminary observational scale can be used before a
professional diagnosis is available.

| Level | Meaning |
|---:|---|
| 0 | No visible concern |
| 1 | Minor or cosmetic |
| 2 | Moderate |
| 3 | Substantial |
| 4 | Severe |
| 5 | Critical |

Possible interpretations are:

- **0:** no abnormal symptoms observed;
- **1:** small leaf spots or minor feeding;
- **2:** noticeable symptoms or limited branch dieback;
- **3:** large affected area or several dead branches;
- **4:** major crown, trunk, or root concerns;
- **5:** immediate structural concern or rapid decline.

Different parts of the same tree should be scored separately.

For example:

| Tree Part | Score |
|---|---:|
| Leaves | 2 |
| Crown | 1 |
| Branches | 3 |
| Trunk | 1 |
| Root Zone | Unknown |

This prevents a serious branch problem from disappearing
inside one average whole-tree score.

## 20. Individual and Population Measurements

Two different questions must be answered.

### Individual-Level Question

How much of one plant is affected?

### Population-Level Question

How many plants are affected?

These correspond to different measurements.

### Incidence

Incidence measures the proportion of sampled plants that meet
a defined condition.

$$
\hat{p}
=
\frac{x}{n}
$$

Where:

- $x$ = number of affected plants;
- $n$ = total number of plants examined;
- $\hat{p}$ = estimated affected proportion.

The percentage is:

$$
\text{Incidence Percentage}
=
\frac{x}{n}
\times
100
$$

If 18 of 60 sampled trees contain a defined symptom:

$$
\hat{p}
=
\frac{18}{60}
=
0.30
$$

Therefore:

$$
\text{Observed Incidence}
=
30\%
$$

This means that 30 percent met the case definition.

It does not necessarily mean that 30 percent were seriously
ill.
```

```
## 21. Measuring Severity

Severity measures how much of an individual plant or tissue
is affected.

For one leaf:

$$
S_{ij}
=
\frac{
A_{ij,\mathrm{affected}}
}{
A_{ij,\mathrm{total}}
}
$$

Where:

- $S_{ij}$ = severity of leaf $j$ on tree $i$;
- $A_{ij,\mathrm{affected}}$ = affected leaf area;
- $A_{ij,\mathrm{total}}$ = total leaf area.

For several sampled leaves from one tree:

$$
S_i
=
\frac{1}{m_i}
\sum_{j=1}^{m_i}
S_{ij}
$$

Where:

- $m_i$ = number of sampled leaves;
- $S_i$ = mean leaf severity for tree $i$.

A tree may have many affected leaves with very small lesions.

Another tree may have only a few affected leaves with severe
tissue loss.

These conditions should not be treated as equivalent.

The study must keep separate:

$$
\text{Incidence}
\neq
\text{Severity}
\neq
\text{Whole-Plant Impact}
$$
```

```
## 22. Population Sampling

Examining only damaged-looking trees would produce a biased
estimate of population health.

The probability of selection should not depend on whether a
tree already appears diseased.

Ideally:

$$
P(
\text{selected}
\mid
\text{visibly damaged}
)
=
P(
\text{selected}
\mid
\text{apparently healthy}
)
$$

Possible sampling methods include:

### Simple Random Sampling

Trees are selected randomly from a complete inventory.

### Systematic Sampling

Every fixed number of trees is selected.

Examples include:

- every fifth tree;
- one tree every 100 metres;
- the nearest tree to each predetermined point.

### Stratified Random Sampling

Trees are divided into groups before random selection.

Possible groups include:

- species;
- street;
- park;
- land-use type;
- tree size;
- planter type;
- neighbourhood;
- distance from the road.

### Cluster Sampling

Street segments, blocks, or parks are selected first.

Trees within those selected groups are then examined.

### Opportunistic Observation

Interesting cases are recorded whenever encountered.

These records are useful for discovery.

They should not be used alone to estimate population
prevalence.

## 23. Urban Sampling Challenges

Urban trees are not distributed randomly.

They may be planted:

- in rows;
- in groups of one species;
- during the same year;
- in similar soil;
- under one maintenance program;
- along the same road;
- near the same construction project.

Nearby trees may therefore share both biological and

A cluster of symptoms may indicate transmission.

It may also reflect:

- the same species;
- the same age;
- the same soil;
- the same pruning history;
- the same drought exposure;
- the same road-salt exposure.

Spatial clustering does not automatically prove that a
pathogen spread between trees.

## 28. Repeated Observations Through Time

One observation may reveal the current condition.

It may not reveal the sequence of events.

Repeated observations can show whether a symptom is:

- expanding;
- stable;
- recovering;
- recurring;
- seasonal;
- spreading to nearby trees;
- associated with weather;
- associated with maintenance activity.

For severity measured at two times:

$$
\Delta S_i
=
S_i(t_2)
-
S_i(t_1)
$$

Where:

- $S_i(t_1)$ = severity during the first observation;
- $S_i(t_2)$ = severity during the later observation;
- $\Delta S_i$ = change in severity.

If:

$$
\Delta S_i > 0
$$

severity increased.

If:

$$
\Delta S_i < 0
$$

severity decreased.

Repeated measurements should use the same method whenever
possible.

### Examples of Seasonal Disease Patterns

Some diseases are especially recognizable at particular times.

#### Snow Blight

Snow blight is typically observed around snowmelt.

Discoloured conifer needles may occur beneath the former snow
line, and temporary web-like mycelium may be visible.

The resemblance to spider webbing demonstrates why timing,
location, host, and persistence must be recorded.

[Snow blight reference](
https://tidcf.nrcan.gc.ca/en/diseases-caused-by-pathogens/factsheet/324
)

#### Tar Spot

Tar spot begins as yellowish-green areas on maple leaves.

The spots enlarge, become raised, and darken over time.

The dramatic black spots may therefore be easiest to recognize
later in the growing season even though infection occurred
earlier.

The fungus survives in fallen infected leaves and produces
spores during the following spring.

Severe infection and premature leaf fall are possible, but the
condition rarely represents a serious threat to an otherwise
healthy tree.

[Tar spot reference](
https://tidcf.nrcan.gc.ca/en/diseases-caused-by-pathogens/factsheet/12
)

#### Scleroderris Canker

Scleroderris canker affects compatible pine hosts.

The disease begins with browning needles and death of buds on
an affected shoot.

The infection can then progress along the shoot.

A canker forms if the disease reaches the trunk.

The North American and European strains may produce similar
initial symptoms but differ in their typical extent and
severity.

The North American strain is generally restricted to the
lower portion of the stem and is rarely fatal.

The European strain is more virulent, can invade the whole
crown, and may kill affected trees over several years.

Visual observation alone may not identify the strain.

[Scleroderris canker: North American strain](
https://tidcf.nrcan.gc.ca/en/diseases-caused-by-pathogens/factsheet/6
)

[Scleroderris canker: European strain](
https://tidcf.nrcan.gc.ca/en/diseases-caused-by-pathogens/factsheet/7
)

#### Oak Leaf Blister

Oak leaf blister is another visually distinctive foliage
condition that may be considered when observing abnormal oak
leaves.

As with other diseases, the host, leaf pattern, season, and
regional distribution should be checked before assigning the
name.

[Oak leaf blister reference](
https://tidcf.nrcan.gc.ca/en/diseases-caused-by-pathogens/factsheet/13
)

#### Powdery Mildew

Powdery mildew describes a group of diseases caused by
multiple related fungi rather than one disease organism that
infects every host.

A database entry associated with one region or host should not
be interpreted as evidence that all powdery mildew is absent
from other regions.

The precise pathogen may differ among plant species and
locations.

[Powdery mildew reference](
https://tidcf.nrcan.gc.ca/en/diseases-caused-by-pathogens/factsheet/1000072
)

These examples are included to demonstrate diagnostic
methodology.

A separate regional disease inventory would be more
appropriate for comparing common illnesses in Waterloo Region.

## 30. Correlation and Causation

Trees near roads may show more damage than trees in parks.

Possible explanations include:

- road salt;
- soil compaction;
- heat;
- air pollution;
- restricted roots;
- pruning;
- vehicle damage;
- different species;
- different tree ages;
- different irrigation.

A statistical association does not identify the cause by
itself.

In general:

$$
\text{Correlation}
\neq
\text{Causation}
$$

A causal claim becomes stronger when there is evidence of:

- temporal order;
- a biological mechanism;
- comparison groups;
- repeated patterns;
- an intervention;
- exclusion of major alternative explanations.

## 31. Treatment and Management

Treatment should follow diagnosis whenever possible.

The wrong treatment may:

- fail to solve the problem;
- waste resources;
- damage beneficial organisms;
- create contamination;
- delay the correct response;
- increase pathogen spread.

Possible management categories include:

### Monitoring

Appropriate when the issue is uncertain, minor, stable, or
likely to recover.

### Environmental Improvement

Examples include:

- appropriate watering;
- improved drainage;
- reduced compaction;
- root protection;
- corrected mulch depth;
- reduced physical disturbance.

### Sanitation

Examples include:

- removal of infected debris;
- cleaning tools;
- proper disposal of material;
- preventing contaminated soil movement.

### Pruning

Pruning may remove:

- dead branches;
- unsafe branches;
- heavily infected tissue;
- damaged branches.

### Biological Management

Possible approaches include:

- natural predators;
- microbial competition;
- resistant plant varieties;
- ecological management.

### Chemical Treatment

Possible treatments may include fungicides, bactericides, or
insecticides.

Chemical treatment should not be recommended from uncertain
visual identification alone.

### Removal

Removal may be considered when:

- structural risk is unacceptable;
- decline is irreversible;
- transmission risk is high;
- treatment is impractical;
- professional guidance recommends removal.

## 32. Evaluating Treatment Effectiveness

Treatment can be studied as an intervention.

A treated group should ideally be compared with an untreated
or differently managed group.

A simple treatment effect can be represented as:

$$
\text{Treatment Effect}
=
\Delta S_{\mathrm{treated}}
-
\Delta S_{\mathrm{comparison}}
$$

Where:

- $\Delta S_{\mathrm{treated}}$ = change in treated plants;
- $\Delta S_{\mathrm{comparison}}$ = change in comparison plants.

Possible outcomes include:

- reduced symptom expansion;
- fewer new infections;
- improved canopy growth;
- reduced mortality;
- improved structural stability;
- reduced pest abundance;
- improved soil conditions.

Improvement after treatment does not automatically prove that
the treatment caused the improvement.

The plant may have recovered naturally.

## 35. Error and Bias

Field diagnosis may produce two important errors.

### False Positive

A healthy or low-concern condition is incorrectly classified
as serious disease.

Examples include:

- harmless lichen identified as a pathogen;
- natural bark shedding identified as severe damage;
- cosmetic leaf spots identified as whole-tree decline.

### False Negative

A serious condition is incorrectly classified as harmless.

Examples include:

- internal decay overlooked because leaves remain green;
- root damage missed because roots are underground;
- early vascular disease missed before visible decline.

The study may also be influenced by:

- selection bias;
- confirmation bias;
- observer bias;
- detection bias;
- spatial bias;
- seasonal bias.

These can be reduced through:

- predefined sampling;
- standardized field cards;
- repeated observations;
- photographs;
- written case definitions;
- multiple observers;
- preservation of uncertain classifications.

## 38. Core Methodological Principles

This project will follow several core principles.

1. A symptom is not automatically a diagnosis.

2. Visible damage does not always indicate serious decline.

3. Serious conditions are not always visually dramatic.

4. Several causes may act together.

5. The visible organism may be a pathogen, opportunist,
   decomposer, or harmless associate.

6. Fungi are natural components of plant ecosystems and are
   not automatically evidence of disease.

7. Seasonal growth, leaf loss, dormancy, and tissue
   replacement must be distinguished from illness.

8. Roots are central to plant health, but not every disease
   begins in the roots.

9. Cultivation and transplanting may create stress without
   proving that a tree is permanently weak.

10. Disease visibility may depend on season, even when the
    original infection occurred earlier.

7. Host species should be identified before assigning a
   disease.

8. Healthy-looking trees must be sampled alongside damaged
   trees.

9. Opportunistic observations must be separated from
   statistical samples.

10. Incidence, severity, and biological impact are different.

11. Leaves, branches, trunks, crowns, and roots should be
    evaluated separately.

12. Repeated observations are needed to measure progression.

13. Spatial clusters do not automatically prove transmission.

14. Correlation does not automatically establish cause.

15. Treatments should follow evidence-based diagnosis.

16. Uncertainty must be recorded rather than hidden.

17. Destructive investigation should not be performed without
    permission and appropriate expertise.

18. Strong conclusions require evidence that can distinguish
    between plausible explanations.

## 39. Conclusion

The analysis of plant illness begins with careful observation,
but it cannot end there.

A complete methodology must connect field observations to:

- host identification;
- symptom description;
- environmental context;
- biological mechanisms;
- competing hypotheses;
- population sampling;
- severity measurements;
- spatial patterns;
- temporal change;
- treatment evaluation.

The central process developed in this notebook is:

$$
\boxed{
\text{Observe}
\rightarrow
\text{Describe}
\rightarrow
\text{Classify}
\rightarrow
\text{Measure}
}
$$

$$
\boxed{
\rightarrow
\text{Compare Hypotheses}
\rightarrow
\text{Collect Evidence}
\rightarrow
\text{Evaluate Seriousness}
}
$$

$$
\boxed{
\rightarrow
\text{Conclude}
\rightarrow
\text{Monitor or Treat}
}
$$

This framework allows unusual possibilities to be
investigated without accepting them prematurely.

It also prevents common or harmless processes from being

The purpose of the project is not merely to create a list of
plant diseases.

Its purpose is to create a systematic framework for deciding:

- what happened;
- what may have caused it;
- how confidently it can be identified;
- how serious it is;
- how common it is;
- whether it is changing;
- what response is justified.
