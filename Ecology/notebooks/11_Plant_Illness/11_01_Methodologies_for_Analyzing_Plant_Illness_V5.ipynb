# 11_01. Methodologies for Analyzing Plant Illness

## From Initial Observation to Evidence, Diagnosis, Population Analysis, and Treatment

Plant illness is rarely explained by one visible symptom.

A damaged leaf, broken branch, patch of missing bark, fungal growth,
insect hole, or declining crown may have many possible causes. Some
symptoms indicate serious threats to survival or structural stability,
while others are temporary, cosmetic, seasonal, or part of ordinary
ecological interactions.

This notebook develops a general methodology for investigating plant
illness. It is intended to show how an observer can move from an initial
observation to organized evidence, competing explanations, measurements,
and a defensible conclusion.

The central question is:

$$
\text{How do we move from an initial observation to a defensible conclusion?}
$$

The investigation will consider:

- how plant illness and damage should be defined;
- why host species and life stage must be identified;
- how plants resist infection and recover from injury;
- biological, environmental, chemical, mechanical, and management causes;
- how symptoms and signs can be observed systematically;
- how competing explanations can be compared;
- how seriousness can be evaluated;
- how individual plants, populations, and communities can be sampled;
- how incidence, severity, and whole-plant impact can be measured;
- how spatial, seasonal, and long-term patterns can be analyzed;
- how prevention, treatment, and management can be evaluated;
- what conclusions are possible with limited equipment, time, and expertise.

This introductory notebook does not attempt to diagnose one particular
tree or conduct a complete field study. Instead, it establishes the
conceptual and methodological foundation for later notebooks involving
field observations, municipal inventories, maps, photographs, statistical
analysis, disease references, and possible management strategies.

## 1. Purpose of the Study

The purpose of this project is to develop a repeatable method for studying
plant illness in urban environments.

The project begins with observations such as:

- dark spots on leaves;
- missing leaf tissue;
- discoloured foliage;
- broken or hanging branches;
- dead branches;
- missing bark;
- wounds on trunks or roots;
- exposed or severed roots;
- unusual fungal structures;
- web-like material;
- insect holes;
- declining or sparse crowns;
- new shoots growing from stressed trunks or branches.

These observations may suggest that a plant is unhealthy, but they do not
automatically identify the cause.

The methodology distinguishes between four stages:

$$
\text{Observation}
\rightarrow
\text{Description}
\rightarrow
\text{Analysis}
\rightarrow
\text{Conclusion}
$$

An **observation** records what is visible.

A **description** converts the observation into measurable characteristics.

An **analysis** compares possible explanations using additional evidence.

A **conclusion** states what the evidence supports, including uncertainty.

The goal is not merely to name diseases. The goal is to understand:

- what happened;
- which causes are plausible;
- how serious the condition is;
- how common it is within and across species;
- whether it is spreading or changing;
- what evidence supports each explanation;
- what additional evidence is needed;
- what preventive or corrective action is justified.

The broader ecological purpose is stewardship: maintaining healthy plant
populations, communities, and ecosystems while responding proportionally
to agents that create serious harm. Monitoring disease is therefore both a
form of care and a form of ecological defence. The objective is not to
eliminate every fungus, bacterium, insect, or competitor. Most are natural
parts of ecosystems. The objective is to recognize when an interaction has
crossed from ordinary ecological activity into severe decline, structural
risk, epidemic spread, or loss of vulnerable species.

## 2. Observation Is Not Diagnosis

A visible symptom is evidence that something has occurred, but it is not necessarily evidence of one specific cause.

For example, a hanging branch might have resulted from:

- wind;
- ice;
- mechanical impact;
- poor attachment;
- internal decay;
- pruning;
- vandalism;
- animal activity;
- a previous infection;
- several interacting causes.

Similarly, a hole in a leaf might have resulted from:

- insect feeding;
- physical tearing;
- hail;
- abrasion;
- dead tissue falling away;
- a bacterial lesion;
- a fungal lesion;
- ordinary accumulated damage.

A responsible investigation must avoid moving directly from:

$$
\text{I observed damage}
$$

to:

$$
\text{I know what caused the damage}
$$

Instead, the appropriate sequence is:

$$
\boxed{
\text{Observed condition}
\rightarrow
\text{possible explanations}
\rightarrow
\text{additional evidence}
\rightarrow
\text{provisional diagnosis}
}
$$

The word **provisional** is important.

A field diagnosis may remain uncertain unless it is supported by expert inspection, microscopy, culturing, DNA analysis, soil testing, chemical analysis, or other specialized methods.

## 3. Signs, Symptoms, Causes, and Diagnoses

Several terms should be kept separate throughout the project.

### Symptom

A symptom is a response of the plant.

Examples include:

- yellowing;
- wilting;
- leaf spots;
- tissue death;
- early leaf loss;
- crown thinning;
- branch dieback;
- cracking;
- swelling;
- reduced growth.

### Sign

A sign is direct physical evidence of a biological agent.

Examples may include:

- fungal fruiting bodies;
- spores;
- visible mycelium;
- bacterial ooze;
- insect larvae;
- eggs;
- feeding galleries;
- shed insect skins;
- frass.

### Injury

An injury is physical or chemical damage that may not involve a living pathogen.

Examples include:

- broken branches;
- cuts;
- girdling;
- salt injury;
- heat injury;
- frost injury;
- root severing;
- herbicide exposure;
- vehicle impact.

### Cause

A cause is the process or agent responsible for the condition.

### Diagnosis

A diagnosis is a reasoned identification of the cause based on the available evidence.

A diagnosis should explain more than the appearance of the plant. It should also be consistent with:

- the host species;
- the affected tissue;
- the symptom pattern;
- the season;
- the location;
- the environmental conditions;
- the likely transmission pathway;
- the progression of the condition over time.

## 4. What Does Plant Illness Mean?

A plant may be considered unhealthy when its ability to survive, grow,
reproduce, maintain its structure, or perform normal physiological functions
has been reduced.

Visible damage does not always produce a meaningful reduction in overall
health. Plants naturally experience cycles of growth, dormancy, leaf
production, leaf loss, flowering, fruiting, tissue death, and tissue
replacement.

A deciduous tree losing leaves in autumn is not ill. A tree may also lose
leaves, twigs, or small branches during stress and still survive.

Visible tissue loss must therefore be interpreted according to:

- species;
- age and life stage;
- season and developmental stage;
- tissue affected;
- amount of tissue affected;
- rate of change;
- ability to replace the tissue.

For example:

- a tree may lose some leaves and remain healthy;
- a leaf disease may appear dramatic but remain mostly cosmetic;
- a tree may replace damaged foliage;
- one branch may die while the rest of the tree survives;
- root disease may be serious before leaf symptoms appear;
- internal decay may create risk without immediately killing the tree;
- seasonal dormancy may resemble decline to an inexperienced observer.

Plants are often highly resilient. Woody plants continue producing new
leaves, roots, shoots, branches, and protective tissues throughout their
lives. This continuing growth allows them to replace tissue, activate dormant
buds, redirect growth, isolate some injuries, and survive partial loss.

Resilience is not unlimited. A plant may approach a threshold beyond which
damage, infection, or resource loss can no longer be compensated for.

Conceptually:

$$
\text{Plant Condition}
=
\text{Damage}
-
\text{Defence and Recovery Capacity}
$$

Serious decline becomes more likely when:

$$
\text{Damage Rate}
>
\text{Recovery Rate}
$$

The threshold differs among species and tissues. Repeated loss of deciduous
leaves may be survivable because a new canopy can be produced later. Many
conifers retain needles for several years and may replace severe foliage loss
more slowly.

Plant health should therefore be treated as a multidimensional condition.

A conceptual model is:

$$
H_i
=
f(
L_i,
C_i,
B_i,
T_i,
R_i,
G_i,
P_i,
E_i,
S_i,
A_i
)
$$

Where:

- $H_i$ = overall condition of plant $i$;
- $L_i$ = leaf condition;
- $C_i$ = crown condition;
- $B_i$ = branch condition;
- $T_i$ = trunk or stem condition;
- $R_i$ = root condition;
- $G_i$ = growth and regeneration;
- $P_i$ = pests or pathogens;
- $E_i$ = environmental conditions;
- $S_i$ = species or genetic characteristics;
- $A_i$ = age or life stage.

This is a conceptual relationship rather than a validated health equation.
It shows that whole-plant health cannot be determined from one damaged leaf,
one visible fungus, or one measurement.

## 5. The Plant Disease Triangle

A traditional model of plant disease considers three interacting components:

1. a susceptible host;
2. a harmful agent;
3. environmental conditions that support disease.

This relationship can be written as:

$$
\text{Disease}
=
f(
\text{Host},
\text{Agent},
\text{Environment}
)
$$

For an urban tree study, this model can be expanded:

$$
D_i(t)
=
f(
H_i,
A_i,
E_i(t),
W_i(t),
V_i(t),
M_i(t)
)
$$

Where:

- $D_i(t)$ = disease or damage outcome for plant $i$ at time $t$;
- $H_i$ = host characteristics;
- $A_i$ = biological or non-biological agent;
- $E_i(t)$ = environmental conditions;
- $W_i(t)$ = wounds and structural damage;
- $V_i(t)$ = vectors and transmission pathways;
- $M_i(t)$ = maintenance and management history.

This expanded model recognizes that an urban tree may be influenced by planting conditions, road salt, pruning, root restriction, construction, irrigation, drought, insects, fungi, bacteria, and human activity at the same time.

## 6. Plant Barriers and Defence Mechanisms

Plants are not passive surfaces on which organisms grow.

They possess physical, chemical, physiological, and ecological barriers that
may prevent infection or limit damage.

### Physical Barriers

Physical barriers include:

- bark;
- waxy leaf cuticles;
- epidermal tissue;
- cell walls;
- protective bud scales;
- sealed tissue surfaces.

### Chemical and Physiological Defences

Plants may produce compounds or responses that:

- inhibit fungal or bacterial growth;
- discourage insect feeding;
- damage invading organisms;
- signal nearby cells;
- reinforce cell walls;
- isolate infected tissue;
- redirect resources toward defence.

### Localized Defence

Plants may isolate or sacrifice a small area to limit spread.

Possible responses include:

- localized cell death;
- formation of barriers around damaged wood;
- compartmentalization of decay;
- shedding infected leaves;
- abandoning damaged twigs or branches.

### Regrowth and Compensation

Plants may compensate for tissue loss through:

- new leaf production;
- dormant bud activation;
- new shoot growth;
- increased growth in surviving tissue;
- redistribution of stored resources;
- replacement of damaged roots or branches.

New growth does not automatically prove that a plant is responding to disease.
It may also be ordinary seasonal growth.

### Wound Closure and Compartmentalization

Trees do not heal wounds in exactly the same way that animals repair injured
tissue. Damaged wood generally remains damaged. A tree may instead form
barriers around the injury, limit the movement of decay, produce new wood
around the wound, and grow callus tissue inward from the edges.

A wound may eventually become covered without the original damaged wood being
restored. Success depends on wound size, depth, location, species, age, tree
vigour, moisture, continued infection, and the amount of living tissue around
the injury.

Small wounds may close successfully. Large wounds, repeated injuries,
cavities, and wounds containing active decay may remain open or structurally
weak.

### Microbial Communities and Defence

Plants interact continuously with communities of fungi, bacteria, and other
microorganisms on their surfaces, inside some tissues, and throughout the soil
around their roots.

Some microorganisms cause disease. Others may:

- compete with pathogens for space and nutrients;
- occupy possible infection sites;
- produce compounds that inhibit pathogens;
- assist nutrient cycling and uptake;
- stimulate plant defence responses;
- improve tolerance to environmental stress.

Mycorrhizal fungi form especially important associations with many plant roots.
They may increase access to water and mineral nutrients in exchange for sugars
produced by the plant.

The presence of fungi or bacteria is therefore not automatically evidence of
illness. A microorganism may be:

- beneficial;
- neutral;
- an endophyte;
- a decomposer;
- an opportunist;
- a secondary pathogen;
- a primary pathogen.

Fungi cause a large proportion of described plant diseases partly because many
species can disperse widely, grow through plant material as hyphae, and produce
enzymes that break down structural compounds. Bacteria can also cause serious
leaf spots, blights, cankers, galls, soft rots, and vascular diseases.

The apparent dominance of fungi in a particular database may reflect both real
biological diversity and the purpose, hosts, geography, and detection methods
represented by that database.

## 7. How Plant Barriers Are Bypassed

Pathogens and pests may reach plant tissue through several routes.

Possible entry pathways include:

- natural openings in leaves;
- young or developing tissue;
- damaged roots;
- broken branches;
- pruning wounds;
- cracks in bark;
- insect feeding wounds;
- frost cracks;
- storm damage;
- mechanical injuries;
- connected infected roots;
- contaminated soil or water;
- infected plant material;
- contaminated tools.

The presence of a wound may increase the chance of infection, but a wound
alone does not prove that infection occurred. Disease near a wound also does
not prove that the wound happened first.

Several sequences are possible:

$$
\text{Wound}
\rightarrow
\text{Infection}
\rightarrow
\text{Decline}
$$

$$
\text{Infection}
\rightarrow
\text{Weakened Tissue}
\rightarrow
\text{Branch Failure}
$$

$$
\text{Environmental Stress}
\rightarrow
\text{Reduced Defence}
\rightarrow
\text{Infection}
\rightarrow
\text{Structural Damage}
$$

A wound creates an opportunity, but infection still depends on whether an
appropriate pathogen reaches the wound, can infect that host, encounters
favourable conditions, and overcomes plant defence and microbial competition.

Therefore:

$$
\text{Wound}
\neq
\text{Infection}
$$

and:

$$
\text{Fungus Near a Wound}
\neq
\text{Cause of the Wound}
$$

Determining the order of events may require repeated observations, internal
inspection, maintenance records, weather data, or professional assessment.

## 8. Major Categories of Plant Illness and Damage

Plant illness should first be classified into broad causal categories. This
prevents the study from assuming that every abnormality is an infectious
disease.

### 8.1 Infectious Pathogens

Potential pathogens include:

- fungi and fungus-like organisms;
- bacteria;
- viruses and viroids;
- phytoplasmas;
- nematodes;
- parasitic plants.

### 8.2 Insects, Mites, and Other Animals

Possible animal-related causes include:

- leaf-chewing insects;
- sap-feeding insects;
- bark beetles;
- wood-boring insects;
- leaf miners;
- gall-forming organisms;
- mites;
- caterpillars;
- vertebrate browsing;
- bark stripping;
- nesting or burrowing activity.

Some organisms directly damage plants. Others act as vectors that transport
pathogens between hosts.

### 8.3 Physical Environmental Stress

Examples include:

- heat;
- drought;
- flooding;
- frost;
- ice;
- wind;
- excessive or insufficient light;
- poor drainage;
- low oxygen around roots.

### 8.4 Chemical and Nutritional Stress

Examples include:

- nutrient deficiency;
- nutrient toxicity;
- excessive fertilizer;
- road salt;
- herbicide exposure;
- air pollution;
- contaminated soil or water;
- unsuitable soil pH.

### 8.5 Mechanical Injury

Examples include:

- storms;
- vehicles;
- construction equipment;
- lawn-maintenance equipment;
- pruning;
- staking materials;
- wires and signs;
- drilling or cutting;
- root excavation;
- pedestrian pressure;
- vandalism.

### 8.6 Cultivation, Planting, and Management

Examples include:

- damaged nursery stock;
- transplant shock;
- root loss during transport;
- incorrect planting depth;
- restricted planting pits;
- circling or girdling roots;
- compacted soil;
- unsuitable species selection;
- excessive or insufficient mulch;
- inadequate or excessive watering;
- repeated disturbance;
- poor pruning technique.

### 8.7 Normal or Low-Concern Processes

Examples include:

- seasonal leaf loss;
- ordinary bark shedding;
- minor insect feeding;
- old pruning wounds;
- harmless lichen or moss;
- branch self-pruning;
- decomposition of tissue that was already dead.

Several categories may operate together. Classification is a starting point,
not a claim that only one cause exists.

### Urban Transplanting and Root Restriction

Urban trees often begin life under conditions that differ from those of trees
that establish naturally in place.

Trees may be transported as:

- container-grown stock;
- bare-root stock;
- balled-and-burlapped stock;
- field-grown nursery stock.

Transportation and planting may disturb or reduce the root system. Root pruning
can sometimes be used deliberately to create a compact transplantable root
ball, but a newly planted tree may still have much less absorbing root area
than an established tree of similar crown size.

Possible consequences include:

- transplant shock;
- reduced water and nutrient uptake;
- slower growth;
- crown dieback;
- reduced stored resources;
- increased drought sensitivity;
- reduced capacity to invest in defence and repair.

After planting, root growth may be restricted by pavement, small soil volumes,
compaction, buried utilities, low soil oxygen, poor drainage, circling roots,
road salt, excavation, and repeated disturbance.

Roots also exchange chemical signals and interact with fungi, bacteria, and
neighbouring plants. Restricted or disturbed rooting conditions may therefore
alter below-ground ecological relationships as well as water and nutrient
acquisition.

A defensible causal sequence is:

$$
\text{Root Loss or Restriction}
\rightarrow
\text{Water and Nutrient Stress}
\rightarrow
\text{Reduced Growth and Defence}
\rightarrow
\text{Greater Susceptibility}
$$

This does not mean that every cultivated tree is permanently weak or that every
urban illness begins in the roots. It identifies one important pathway that
should be investigated.

## 9. Primary, Secondary, and Opportunistic Agents

The organism most visible on a damaged tree may not be the original cause of decline.

### Primary Agent

A primary agent can attack relatively healthy tissue and initiate the condition.

### Secondary Agent

A secondary agent attacks tissue that has already been weakened by another stressor.

### Opportunistic Organism

An opportunistic organism takes advantage of damaged, dying, or dead tissue.

### Decomposer

A decomposer breaks down tissue after it has died.

A possible sequence is:

$$
\text{Drought}
\rightarrow
\text{reduced plant defence}
\rightarrow
\text{insect attack}
\rightarrow
\text{branch death}
\rightarrow
\text{fungal decomposition}
$$

In this example, the visible fungus might not have caused the original decline.

The methodology must therefore ask:

- Did the organism attack healthy tissue?
- Was the tissue already injured?
- Was the plant already stressed?
- Is the organism spreading into living tissue?
- Is the organism restricted to dead material?
- Is the visible organism a pathogen, an opportunist, or a decomposer?

## 10. Transmission Pathways and Vectors

A vector is an organism that carries a pathogen from one host to another.

Possible vectors include:

- beetles;
- aphids;
- leafhoppers;
- mites;
- nematodes;
- birds;
- mammals;
- humans.

Not every transmission pathway requires a living vector.

Other pathways include:

- wind;
- rain splash;
- surface water;
- connected roots;
- soil movement;
- contaminated tools;
- nursery stock;
- transported firewood;
- fallen leaves;
- infected seeds;
- movement of branches or wood debris.

Host species can alter transmission. One species may be highly susceptible,
another may become infected without severe symptoms, and another may resist
infection. A tolerant host may still carry an agent and contribute to spread,
while a resistant or non-host species may interrupt transmission and act as a
partial ecological buffer.

Field observations should record both:

- **Suspected Agent**
- **Suspected Transmission or Damage Pathway**

This prevents wind, water, or tools from being incorrectly classified as
biological vectors.

## 11. Types of Plant Disease by Affected Tissue

Plant diseases and injuries can be organized by the tissue or system affected.

### Foliage Conditions

These may involve leaf spots, blights, mildew, rusts, galls, chlorosis,
necrosis, premature leaf loss, holes, skeletonization, or leaf mines.

### Shoot and Twig Conditions

These may involve shoot blight, dead buds, distorted growth, twig dieback,
witches' broom, or stunted shoots.

### Branch Conditions

These may involve branch dieback, cankers, fractures, decay, weak unions,
hanging branches, bark loss, or pruning wounds.

### Trunk and Stem Conditions

These may involve cankers, cavities, cracking, missing bark, decay, galleries,
fungal fruiting bodies, bleeding, oozing, or girdling.

### Vascular Conditions

These may interfere with water or nutrient transport and may produce wilting,
rapid crown decline, discoloured vascular tissue, one-sided decline, or reduced
growth.

### Root and Root-Collar Conditions

These may involve root decay, severing, girdling roots, compacted soil, poor
drainage, excavation damage, exposed roots, or root-collar burial.

Roots are sometimes compared with a digestive system because they are central
to acquiring water and mineral nutrients. The analogy is useful but incomplete.
Roots also anchor the plant, store carbohydrates, exchange gases, produce and
respond to chemical signals, and interact with fungi, bacteria, and neighbouring
roots.

Many serious problems begin in or around the roots, including root rot,
waterlogging, drought, nutrient imbalance, nematodes, root-feeding insects,
girdling roots, severed roots, chemical contamination, and planting too
deeply.

Root problems may later produce above-ground symptoms such as small leaves,
yellowing, reduced growth, crown thinning, branch dieback, or instability.

However, not every disease begins in the roots. Leaves, flowers, buds, shoots,
branches, bark, wounds, and vascular tissues may also be attacked directly.

A useful diagnostic question is:

$$
\text{Could the visible above-ground symptom reflect an unseen root problem?}
$$

## 12. A Diagnostic Question Tree

The diagnostic process can be approached like a dichotomous key. It does not
immediately identify one disease. Instead, it narrows possible causal
categories.

### Step 1: Identify the Host

Determine the plant species, or at minimum the genus or broader group.

Ask:

- Which species is this?
- How confident is the identification?
- What age or life stage is it?
- Is it a cultivar or planted variety?
- Is the suspected condition known to affect this host?

Host identification is essential because species differ in:

- susceptibility and resistance;
- visible symptoms;
- seasonal growth;
- nutrient requirements;
- bark and leaf structure;
- root architecture;
- ability to compartmentalize wounds;
- relationships with vectors and microbes.

A disease may be visually distinctive on one species and subtle on another.
A conspicuous host may therefore help reveal an agent that is less obvious on
other nearby species.

### Step 2: Decide Whether the Feature Is Abnormal

Ask:

- Is it present on healthy plants of the same species?
- Is it normal for the season or developmental stage?
- Is it ordinary bark texture, leaf shape, or growth?
- Is it old damage rather than active damage?
- Is it lichen, moss, or another low-concern organism?

### Step 3: Identify the Tissue Affected

Choose the primary location:

- leaves;
- buds;
- shoots;
- branches;
- trunk or bark;
- vascular tissue;
- roots or root collar;
- flowers or fruit;
- the entire plant.

### Step 4: Describe Distribution Within the Plant

Determine whether the condition affects:

- one leaf;
- several leaves;
- one branch;
- one side of the crown;
- the lower or upper crown;
- the trunk base;
- the entire plant.

### Step 5: Describe Distribution Among Plants

Determine whether it affects:

- one individual;
- several individuals of one species;
- several species;
- one street or park;
- several separated areas;
- an entire site.

### Step 6: Look for Evidence of a Living Agent

Look for fungal structures, spores, possible mycelium, insects, eggs, larvae,
galleries, frass, bacterial ooze, or repeated feeding patterns.

### Step 7: Look for Physical, Chemical, or Environmental Injury

Look for clean cuts, jagged breaks, torn fibres, impact marks, severed roots,
machinery wounds, ties, construction disturbance, heat, drought, flooding,
road salt, fertilizer injury, pollution, or poor drainage.

### Step 8: Examine Spatial and Seasonal Patterns

Ask whether the condition follows one species, planting age, road, drainage
area, pruning zone, construction area, weather event, connected root system,
or particular time of the growing season.

### Step 9: Determine Whether the Condition Is Progressing

Repeated observations should determine whether it is stable, spreading,
recovering, recurring seasonally, or appearing on additional plants.

## 13. Determining Whether an Issue Is Serious

The visual size of a symptom is not always equal to its biological seriousness.

A condition should be evaluated using several dimensions.

### 13.1 Tissue Importance

Damage to different tissues has different consequences.

Leaf damage may reduce photosynthesis but may be temporary.

Root, vascular, trunk, and structural damage may have greater long-term consequences.

### 13.2 Extent

The amount of tissue affected should be estimated.

Examples include:

- percentage of leaf area damaged;
- percentage of leaves affected;
- percentage of crown affected;
- percentage of branches dead;
- percentage of trunk circumference damaged;
- number of exposed or severed roots.

### 13.3 Rate of Progression

A rapidly expanding condition may require more concern than stable old damage.

### 13.4 Structural Risk

A tree may remain alive while becoming structurally dangerous.

Potential concerns include:

- large hanging branches;
- major cavities;
- extensive decay;
- root instability;
- trunk cracks;
- weak branch unions;
- substantial bark loss around the trunk.

### 13.5 Host Importance

The consequences may depend on whether the plant is:

- rare;
- old;
- culturally important;
- part of a protected population;
- a major shade tree;
- near buildings or walkways;
- one of many replaceable ornamental plants.

### 13.6 Population Spread

A condition affecting one tree may be less urgent than one rapidly appearing across many trees.

### 13.7 Reversibility

The investigation should ask whether the condition is:

- naturally recoverable;
- manageable;
- treatable;
- preventable;
- irreversible;
- likely to recur.

## 14. A Provisional Seriousness Framework

A preliminary seriousness classification can be used before a professional diagnosis is available.

| Level | Interpretation | Example Characteristics |
|---|---|---|
| 0 | No visible concern | No abnormal symptoms observed |
| 1 | Minor or cosmetic | Small leaf spots, minor feeding, old wound |
| 2 | Moderate | Repeated symptoms, limited branch dieback, noticeable defoliation |
| 3 | Substantial | Large affected area, multiple dead branches, expanding damage |
| 4 | Severe | Major crown decline, large wounds, root or trunk concerns |
| 5 | Critical | Immediate structural concern, rapid decline, widespread mortality |

This classification is observational.

It should not be presented as a validated medical or arboricultural scale unless it is later connected to an established assessment system.

Different parts of the same tree should be scored separately.

For example:

| Tree Component | Score |
|---|---:|
| Leaves | 2 |
| Crown | 1 |
| Branches | 3 |
| Trunk | 1 |
| Root Zone | Unknown |

This prevents a serious branch hazard from being hidden inside an average whole-tree score.

## 15. Building Competing Hypotheses

Every important observation should generate multiple plausible explanations.

For example, consider a partially detached branch.

Possible hypotheses include:

1. the branch was broken by wind;
2. the branch was damaged by ice;
3. the branch was struck by a person or object;
4. the branch failed because of a weak attachment;
5. decay weakened the branch before it broke;
6. the branch was partly cut during maintenance;
7. improper pruning left the branch attached;
8. an animal damaged the branch;
9. the branch was intentionally damaged;
10. several causes acted together.

The purpose of the investigation is not to list possibilities forever.

The purpose is to identify evidence that distinguishes them.

| Hypothesis | Supporting Evidence | Contradicting Evidence | Evidence Still Needed |
|---|---|---|---|
| Wind breakage | Jagged fibres, recent wind event | Smooth tool marks | Weather date, close photographs |
| Pruning | Repeated cuts, municipal work | Random tear pattern | Maintenance records |
| Internal decay | Soft or discoloured wood | Healthy internal tissue | Arborist inspection |
| Mechanical impact | Scraping or impact pattern | No external injury | Site history |
| Intentional damage | Repeated targeted marks | Ordinary maintenance pattern | Strong physical or documentary evidence |

Unusual explanations should remain possible when appropriate, but they should not receive greater weight than the evidence supports.

## 16. Evidence and Confidence

Not all evidence has the same strength.

### Low-Strength Evidence

Examples include:

- resemblance to an online photograph;
- one observation;
- an uncertain memory;
- a symptom shared by many diseases;
- assumptions about what happened before arrival.

### Moderate-Strength Evidence

Examples include:

- repeated observations;
- characteristic symptom patterns;
- confirmed host species;
- appropriate seasonal timing;
- similar symptoms on nearby plants;
- weather or maintenance records.

### High-Strength Evidence

Examples include:

- characteristic signs of the causal agent;
- professional arborist assessment;
- microscopy;
- laboratory culturing;
- molecular identification;
- chemical testing;
- internal structural assessment;
- confirmed municipal maintenance records.

A confidence field can accompany every diagnosis:

- **Low confidence**
- **Moderate confidence**
- **High confidence**
- **Confirmed**

The absence of certainty is not a failure.

Recording uncertainty is part of the scientific result.

### Accessible Reference Tools

Careful qualitative observation can narrow possibilities substantially when it
is combined with established references. Useful low-cost tools include a camera,
hand lens, ruler, tree-identification guide, glossary, symptom index, maps,
weather records, municipal inventories, and repeated visits.

The Government of Canada's database provides searchable information on trees,
insects, mites, and diseases:

- [Trees, insects, mites, and diseases of Canada's forests](https://tidcf.nrcan.gc.ca/en)
- [Diseases caused by pathogens](https://tidcf.nrcan.gc.ca/en/diseases-caused-by-pathogens?index=a#index)

A reference can generate or narrow hypotheses, but it does not replace host
identification, geographic range, tissue pattern, seasonal timing, and—when
needed—expert or laboratory confirmation.

Photographs should be collected systematically rather than only as close-up images of the most dramatic symptom.

For each tree, useful photographs may include:

3. the full crown;
4. the trunk;
5. the root collar;

7. the affected branch;
8. the upper and lower leaf surfaces;
9. a close-up of signs or symptoms;
10. a scale reference;
11. nearby trees for comparison;
12. the tree tag or inventory number.

- How much of the tree is affected?
- Is the damage isolated or repeated?

## 19. Tree and Host Identification

Host identification should occur before a named disease is assigned whenever
possible.

Many pests and pathogens are limited to particular hosts, while others affect
related species differently. The same agent may produce severe symptoms in one
species, mild symptoms in another, and no visible symptoms in a third.

Useful identification features include:

- leaf shape and arrangement;
- bark;
- buds and twigs;
- flowers and fruit;
- branching pattern;
- overall form;
- municipal tree tags;
- planting records.

Species uncertainty should be recorded explicitly:

- confirmed species;
- probable species;
- genus only;
- broad group only;
- unknown.

Host identification contributes to diagnosis in several ways.

### Host Range

A suspected disease should be biologically compatible with the host.

### Symptom Expression

Different species may display different symptoms after exposure to the same
agent.

### Nutritional and Environmental Requirements

Species differ in water demand, preferred soil, pH tolerance, light, nutrient
requirements, frost tolerance, salt tolerance, and root architecture.

A condition that appears infectious may instead reflect a species planted in
an unsuitable environment.

### Comparative Detection

A disease may be easy to recognize on one species and difficult to see on
another. A conspicuous symptom on a highly responsive host may provide a clue
to examine less visibly affected hosts nearby.

### Susceptibility, Tolerance, and Resistance

A susceptible species may develop severe symptoms. A tolerant species may
carry the agent with limited decline. A resistant or non-host species may
remain unaffected and may reduce opportunities for transmission.

An incorrect host identification can therefore lead to an incorrect diagnosis,
incorrect severity estimate, and inappropriate treatment.

## 20. Individual, Population, Demographic, and Community Measurements

Plant illness can be studied at several nested ecological levels.

These levels should not be treated as interchangeable because each answers a
different kind of question.

### Individual Level

The individual level examines the condition of one plant.

Questions include:

- How much of one plant is affected?
- Which tissues are affected?
- Is the condition progressing?
- Is the plant compensating or declining?
- Does its response differ from nearby individuals?

Possible measurements include leaf damage, crown dieback, branch condition,
trunk wounds, root-zone stress, growth, regeneration, and symptom progression.

### Study Population

For practical municipal analysis, all trees inside a defined study boundary may
be treated as one large **study population**.

For example:

$$
\mathcal{P}
=
\{\text{all inventoried public trees in Waterloo}\}
$$

This operational population may contain many species and many demographic
groups.

In strict population ecology, a biological population normally refers to
individuals of the same species within a defined area. The larger municipal
collection is therefore also an ecological community when it contains multiple
species.

### Demographic Subpopulations

The study population can be divided into smaller **demographic
subpopulations** according to variables relevant to the question.

Possible demographics include:

- species;
- genus;
- age or size class;
- young, mature, or declining condition;
- street, park, or neighbourhood;
- planting environment;
- management history;
- exposure class;
- observation season.

For species $s$ and age class $a$, a demographic group may be written as:

$$
\mathcal{P}_{s,a}
=
\{i : \text{tree } i \text{ belongs to species } s
\text{ and age class } a\}
$$

The same tree may belong to several useful groupings, such as maple, mature,
street-planted, and high-road-exposure.

Which demographics should be recorded depends on the study question.

If the question concerns species differences, species must remain explicit.
If the question concerns age, age classes must be sampled and indexed. If a
variable is not being studied and is adequately controlled or approximately
constant, it may not need a separate index in that analysis.

### Same-Species Population Level

Within species $s$:

$$
\hat{p}_s
=
\frac{x_s}{n_s}
$$

Where:

- $x_s$ = affected sampled trees of species $s$;
- $n_s$ = total sampled trees of species $s$;
- $\hat{p}_s$ = estimated incidence for species $s$.

A more specific demographic estimate can include age class $a$ and location
$l$:

$$
\hat{p}_{s,a,l}
=
\frac{x_{s,a,l}}{n_{s,a,l}}
$$

This makes it possible to compare like with like rather than combining young
park maples, mature street maples, and newly transplanted maples into one value.

### Community Level

A plant community contains multiple species populations living in the same
area. Community ecology compares those populations and examines how species
traits, abundance, spacing, and interactions affect the condition of the
larger system.

Questions include:

- How many species are affected?
- What proportion of each species is affected?
- Which species show the greatest severity or fastest progression?
- Are some species infected but only mildly affected?
- Can some species carry or transmit an agent while showing few symptoms?
- Do resistant or non-host species interrupt pathways between susceptible
  hosts?
- Do susceptible species form connected pathways that facilitate spread?
- Does species diversity increase community resilience?
- Which species or demographic groups require the greatest conservation
  attention?

The proportion of observed species with at least one defined case is:

$$
\hat{q}
=
\frac{k_{\mathrm{affected}}}{k_{\mathrm{observed}}}
$$

Where:

- $k_{\mathrm{affected}}$ = number of species with at least one case;
- $k_{\mathrm{observed}}$ = total number of species examined.

A cross-species incidence ratio is:

$$
R_{s_1,s_2}
=
\frac{\hat{p}_{s_1}}{\hat{p}_{s_2}}
$$

If $R_{s_1,s_2}>1$, species $s_1$ has a higher observed incidence than species
$s_2$ in the sample.

That difference might reflect susceptibility, tolerance, symptom visibility,
age structure, planting density, environmental exposure, or a transmission
mechanism. It is evidence of a pattern, not an automatic causal explanation.

### Comparing Progression Across Species

For species $s$:

$$
g_s
=
\frac{1}{n_s}
\sum_{i=1}^{n_s}
\frac{S_{i,s}(t_2)-S_{i,s}(t_1)}{t_2-t_1}
$$

Where:

- $S_{i,s}(t)$ = severity for individual $i$ of species $s$ at time $t$;
- $g_s$ = mean observed severity-change rate for species $s$.

If a condition progresses slowly in one species but rapidly in another under
similar conditions, possible explanations include different susceptibility,
defence, recovery, vector preference, symptom visibility, or environmental
exposure.

### Buffers, Bridges, and Community Structure

Species may act as susceptible hosts, tolerant hosts, reservoirs, bridge hosts,
non-hosts, resistant hosts, or possible buffer species.

These roles must be demonstrated rather than assumed. A species with few visible
symptoms might be resistant, but it might also carry an agent without obvious
damage.

Community-level evidence can guide surveillance, conservation priorities,
planting diversity, spacing, replacement decisions, and containment.

A whole-city average can still be calculated, but it may conceal major
differences among demographic subpopulations. Species, age, and other important
demographics should therefore remain explicit until there is a justified reason
to aggregate them.

## 21. Measuring Disease Incidence

Disease incidence describes the proportion of sampled individuals that meet a defined condition.

For a sample of trees:

$$
\hat{p}
=
\frac{x}{n}
$$

Where:

- $x$ = number of trees meeting the case definition;
- $n$ = total number of trees examined;
- $\hat{p}$ = estimated proportion affected.

The percentage affected is:

$$
\text{Incidence Percentage}
=
\frac{x}{n}
\times
100
$$

For example, if 18 of 60 sampled maple trees contain tar-spot-like symptoms:

$$
\hat{p}
=
\frac{18}{60}
=
0.30
$$

Therefore:

$$
\text{Observed Incidence}
=
30\%
$$

This does not necessarily mean that 30% of the trees are seriously ill.

It means that 30% met the specific observational case definition used in the survey.

## 22. Measuring Disease Severity

Disease severity describes how much of an individual plant or tissue is
affected.

### Severity of One Leaf

For leaf $j$ on tree $i$:

$$
S_{ij}
=
\frac{A_{ij,\mathrm{affected}}}{A_{ij,\mathrm{total}}}
$$

Where:

- $S_{ij}$ = severity for leaf $j$ on tree $i$;
- $A_{ij,\mathrm{affected}}$ = affected leaf area;
- $A_{ij,\mathrm{total}}$ = total leaf area.

### Mean Leaf Severity for One Tree

For a sample of $m_i$ leaves from tree $i$:

$$
S_i
=
\frac{1}{m_i}
\sum_{j=1}^{m_i}
S_{ij}
$$

Where:

- $m_i$ = number of sampled leaves from tree $i$;
- $S_i$ = mean estimated leaf severity for tree $i$.

A tree may have high incidence among its leaves but low severity per leaf.
Many leaves may contain one small spot, while a few leaves on another tree may
be almost completely damaged. These conditions are not equivalent.

### Mean Severity Across Sampled Trees

If $n$ trees are sampled, average tree-level severity is:

$$
\bar{S}
=
\frac{1}{n}
\sum_{i=1}^{n}
S_i
$$

Where:

- $n$ = number of sampled trees;
- $S_i$ = severity estimate for tree $i$;
- $\bar{S}$ = mean severity across sampled trees.

This population-level average should not replace the distribution of individual
values. Two populations may have the same mean while one has uniform moderate
damage and the other has many healthy trees plus a few severely affected trees.

### Severity by Species or Demographic Group

For species $s$ with $n_s$ sampled trees:

$$
\bar{S}_s
=
\frac{1}{n_s}
\sum_{i=1}^{n_s}
S_{i,s}
$$

For species $s$, age class $a$, and location class $l$:

$$
\bar{S}_{s,a,l}
=
\frac{1}{n_{s,a,l}}
\sum_{i=1}^{n_{s,a,l}}
S_{i,s,a,l}
$$

These indices allow severity to be compared among demographic
subpopulations without losing species, age, or location information.

### Nested Tissue, Tree, Population, and Community Levels

Severity can therefore be aggregated in stages:

$$
\text{Affected Area}
\rightarrow
\text{Leaf Severity}
\rightarrow
\text{Tree Severity}
\rightarrow
\text{Demographic Mean}
\rightarrow
\text{Community Summary}
$$

Each aggregation removes detail. The original tree-level observations should be
preserved even when larger summaries are calculated.

Severity also needs to be interpreted according to tissue. A numerical value
for leaf area loss is not equivalent to the same numerical proportion of trunk
circumference, roots, or major branches.

## 23. Incidence, Severity, and Whole-Tree Health

Three measurements should be kept separate.

### Incidence

How many plants or plant parts are affected?

### Severity

How much of each plant or plant part is affected?

### Health Impact

How much does the condition affect survival, growth, function, or structural stability?

These relationships can be represented as:

$$
\text{Visible Incidence}
\neq
\text{Disease Severity}
\neq
\text{Whole-Plant Impact}
$$

A visually obvious foliage disease may affect many leaves but have a limited effect on the long-term survival of a mature tree.

A less visible root or vascular problem may have a much greater effect.

The affected tissue and biological mechanism must therefore be considered alongside numerical severity.

## 24. Population and Community Sampling

Examining only damaged-looking trees would produce a biased estimate of plant
health.

Ideally:

$$
P(\text{selected}\mid\text{visibly damaged})
=
P(\text{selected}\mid\text{apparently healthy})
$$

### Defining the Study Extent

Before sampling, define:

- the geographic boundary;
- the target tree population;
- whether private, public, street, and park trees are included;
- the observation period;
- the disease or symptom case definition;
- the demographic variables required by the question.

### Demographic Sampling

Community sampling requires deciding which demographic subpopulations must be
represented fairly.

Possible strata include:

- species;
- age or size class;
- location;
- planting environment;
- management class;
- exposure class;
- symptom or host category.

If the study compares species, each selected species needs an adequate sample.
If it compares age, young and mature trees need representation. A citywide
random sample may contain too few members of a rare but important group, so
stratified sampling may be necessary.

A sample allocation can be proportional:

$$
n_h
=
n\frac{N_h}{N}
$$

Where:

- $N_h$ = number of trees in demographic stratum $h$;
- $N$ = total number of trees in the sampling frame;
- $n$ = total desired sample size;
- $n_h$ = proportional sample assigned to stratum $h$.

Rare or high-priority groups may instead be oversampled. If that occurs,
weights may be needed when estimating a whole-community value.

### Possible Sampling Methods

#### Simple Random Sampling

Trees are selected randomly from a complete inventory.

#### Systematic Sampling

Every fixed number of trees or the closest tree to predetermined points is
selected.

#### Stratified Random Sampling

The population is divided by selected demographics, and trees are sampled
within every required group.

#### Cluster Sampling

Street segments, blocks, parks, or neighbourhoods are selected before trees
within them are examined.

#### Complete Local Census

All trees inside selected small areas are recorded. This is useful for studying
local community composition and possible transmission pathways.

#### Opportunistic Observation

Interesting cases are recorded whenever encountered. These observations are
valuable for discovery and diagnosis but should be kept separate from samples
used to estimate prevalence.

The amount of demographic information collected should match the intended
analysis. A narrow study may need only species and age. A reusable citywide
dataset benefits from richer demographic information because later studies can
select the variables they need.

## 25. Urban Sampling Challenges and Variable Isolation

Urban trees are not distributed randomly.

They may be planted in rows, grouped by species, installed during the same
year, placed in similar soil, or managed under the same program.

A cluster may indicate transmission, but it may also reflect species, age,
soil, drought, road salt, construction, pruning, planting density, irrigation,
or symptom visibility.

The more information collected for each tree, the more ways the observations
can later be grouped and compared.

Potential demographic and environmental variables include:

- species;
- age or trunk diameter;
- street, park, or neighbourhood;
- soil and moisture condition;
- shade;
- road distance;
- planting type;
- pruning history;
- symptom class;
- suspected agent;
- surrounding species;
- observation season.

A general indexed observation may be written as:

$$
Y_{i,s,a,l,t}
$$

Where:

- $i$ = individual tree;
- $s$ = species;
- $a$ = age or size class;
- $l$ = location or site class;
- $t$ = observation time.

Additional indices can be added only when they are required by the question.

### Introducing a Statistical Model

A simple model can be written as:

$$
Y_i
=
\beta_0
+
\beta_1X_{i,1}
+
\beta_2X_{i,2}
+
B_i
+
\epsilon_i
$$

Where:

- $Y_i$ = measured or predicted outcome for tree $i$;
- $X_{i,1}$ and $X_{i,2}$ = measured explanatory variables;
- $\beta_0$ = baseline or intercept;
- $\beta_1$ and $\beta_2$ = coefficients estimated from the data, representing
  the conditional association of each included variable with the outcome;
- $B_i$ = an aggregated background contribution from minor, stable, or
  unmeasured influences;
- $\epsilon_i$ = unexplained variation and measurement error.

The coefficients are calibrated from observations rather than chosen by
intuition. This resembles the training or fitting stage of a statistical or
machine-learning model, although artificial intelligence is not required.

### Keeping Important Variables Explicit

A variable should remain explicit when it varies meaningfully, is central to
the question, may confound another relationship, or is needed for a fair
comparison.

Species should usually remain explicit because it can strongly affect
susceptibility, tolerance, symptoms, nutrient requirements, root architecture,
defence, and relationships with vectors or pathogens.

### Aggregating Low-Impact Variables

Several weak or currently unmeasured influences may be grouped as:

$$
B_i
=
f(U_{i,1},U_{i,2},\ldots,U_{i,r})
$$

Aggregation does not prove that those variables are biologically unimportant.
It means they are not being estimated separately in that analysis.

If one background variable later appears important, it can be brought forward:

$$
B_i
=
\beta_3X_{i,3}
+
B_i^{\ast}
$$

The model becomes:

$$
Y_i
=
\beta_0
+
\beta_1X_{i,1}
+
\beta_2X_{i,2}
+
\beta_3X_{i,3}
+
B_i^{\ast}
+
\epsilon_i
$$

A variable can therefore be added, removed, grouped, separated, held
approximately constant, used as an interaction, or used only as a filter.

A focused study does not need measurements of everything. A large reusable
dataset may collect richer demographics so later analyses can let statistical
evidence show which variables matter for each outcome.

## 26. Municipal Tree Inventory Data

Municipal tree inventories may provide information such as:

- tree coordinates;
- species;
- tag number;
- trunk diameter;
- planting location;
- maintenance history;
- condition category;
- ownership;
- inspection date.

Inventory data could be used to:

- map the tree population;
- count trees by species;
- select random samples;
- compare streets or parks;
- connect field observations to known tree IDs;
- examine spatial clusters;
- estimate the population represented by a sample.

However, inventory data may be incomplete.

Possible limitations include:

- missing recently planted trees;
- removed trees remaining in the dataset;
- incorrect coordinates;
- outdated species identifications;
- private trees being excluded;
- incomplete maintenance histories;
- inconsistent condition categories.

Field verification may therefore be necessary.

## 27. Working Without a Complete Inventory

A study can still be conducted if no complete tree inventory is available.

Possible alternatives include:

- dividing streets into fixed-length segments;
- selecting random coordinates;
- selecting random intersections;
- sampling every fixed number of trees;
- creating small quadrats or plots;
- surveying all trees within selected blocks;
- mapping trees manually during fieldwork.

The sampled area must be defined before observations are used to estimate population prevalence.

For example:

$$
\text{Target Population}
=
\text{all public street trees within selected road segments}
$$

This is more defensible than making conclusions about all trees in the city from a small number of convenient observations.

## 28. Repeated Observations, Seasonality, and Phenology

One observation may reveal the current condition but not the sequence of
events.

Plants pass through seasonal developmental stages known as phenological stages,
including bud swelling, leaf emergence, flowering, fruiting, shoot extension,
leaf-colour change, leaf fall, and dormancy.

A structure or symptom may be normal during one stage and unusual during
another. Pathogens and insects also have seasonal life cycles.

A diagnostic record should therefore include:

- observation date;
- host developmental stage;
- recent weather;
- when the symptom first appeared;
- whether it is new or remains from an earlier season;
- whether the suspected agent is expected at that time.

Some diseases are most visible at characteristic times. Snow-associated blights
may be apparent near snowmelt, while tar-like leaf spots may become much more
obvious later in the growing season. A canker may progress from a shoot toward
a branch or stem over time.

The date a symptom becomes visible may differ from the date infection began:

$$
\text{Date Symptoms Are Observed}
\neq
\text{Date Infection Began}
$$

Repeated observations can show whether a symptom is expanding, stable,
recovering, recurring, seasonal, spreading to nearby plants, associated with
weather, or associated with maintenance.

For severity measured at two times:

$$
\Delta S_i
=
S_i(t_2)
-
S_i(t_1)
$$

Positive values indicate increasing measured severity, while negative values
indicate decreasing severity.

Repeated measurements should use the same tree, tissue category, method,
sampling effort, and photograph position whenever possible.

Disease examples and detailed regional identification notes should be kept in a
separate disease inventory so that this section remains focused on temporal
methodology.

## 29. Environmental and Contextual Data

Plant observations may be compared with environmental variables.

Possible data include:

- temperature;
- rainfall;
- drought indices;
- soil moisture;
- humidity;
- wind;
- storm events;
- flooding;
- road-salt application;
- air pollution;
- construction history;
- pruning schedules;
- irrigation;
- soil type;
- canopy cover;
- impervious surface;
- distance from roads;
- distance from buildings.

A conceptual model might be:

$$
Y_i
=
\beta_0
+
\beta_1 T_i
+
\beta_2 M_i
+
\beta_3 C_i
+
\beta_4 R_i
+
\epsilon_i
$$

Where:

- $Y_i$ = observed disease, damage, susceptibility, or severity;
- $T_i$ = temperature exposure;
- $M_i$ = moisture condition;
- $C_i$ = soil compaction or construction disturbance;
- $R_i$ = road-related exposure;
- $\beta_0$ = baseline expected value when the included predictors are at
  their reference or zero values;
- $\beta_1,\beta_2,\beta_3,\beta_4$ = estimated coefficients describing how
  the predicted outcome changes with each variable while the other included
  variables are held constant;
- $\epsilon_i$ = unexplained variation and measurement error.

For example, $\beta_1$ represents the estimated change in $Y_i$ associated
with a one-unit change in temperature exposure, after accounting for the other
variables included in the model.

The coefficients are estimated from observed data.

They should not be invented or selected only because they appear reasonable.

### A General Variable Form

The model can be written more generally as:

$$
Y_i
=
\beta_0
+
\sum_{j=1}^{p}
\beta_j X_{i,j}
+
\epsilon_i
$$

Where:

- $p$ = number of variables included;
- $X_{i,j}$ = value of variable $j$ for tree $i$;
- $\beta_j$ = estimated contribution of variable $j$.

Changing the study question may change $p$ and the variables included in the
sum.

For one analysis, species and temperature may be central.

For another, age, moisture, and road distance may be more important.

A variable excluded from one model is not being declared biologically
irrelevant.

It is simply not being estimated separately in that model.

### Interaction Effects

Sometimes one variable changes the effect of another.

For example, high temperature may affect a drought-stressed young tree
differently from a well-watered mature tree.

An interaction can be written as:

$$
Y_i
=
\beta_0
+
\beta_1 T_i
+
\beta_2 M_i
+
\beta_3(T_iM_i)
+
\epsilon_i
$$

Where:

- $T_iM_i$ = interaction between temperature and moisture;
- $\beta_3$ = estimated change in the temperature effect associated with
  moisture conditions.

This model should not be interpreted causally without an appropriate study
design.

It is initially a way to investigate conditional patterns and associations.


## 30. Spatial Analysis

Tree observations can be represented as geographic points.

Each point may contain attributes describing:

- species;
- health condition;
- symptom type;
- severity;
- suspected cause;
- observation date;
- confidence;
- photographs;
- nearby environmental conditions.

Spatial analysis may investigate:

- clusters of symptoms;
- patterns by species;
- patterns along roads;
- relationships with parks;
- relationships with construction;
- relationships with drainage;
- relationships with heat exposure;
- relationships with impervious surfaces;
- disease spread through time.

A cluster does not automatically prove biological transmission.

## 33. Prevention, Treatment, and Management Responses

Management should follow diagnosis whenever possible.

The wrong intervention may fail, waste resources, damage beneficial organisms,
create contamination, or increase spread.

### Prevention

Prevention aims to reduce the probability that severe disease begins or
spreads.

Possible approaches include:

- planting suitable species in suitable sites;
- maintaining species and genetic diversity;
- avoiding large continuous plantings of one susceptible host;
- using resistant varieties where appropriate;
- providing adequate soil volume, drainage, and water;
- protecting roots and trunks from construction damage;
- cleaning tools;
- inspecting nursery stock;
- managing infected debris;
- monitoring high-risk sites and seasons;
- spacing susceptible hosts appropriately;
- using resistant or non-host vegetation as partial buffers where evidence
  supports that strategy;
- avoiding management practices that create continuous transmission pathways.

Community data may help identify whether a planting design creates:

- dense clusters of susceptible hosts;
- bridge populations between infected areas;
- resistant gaps or buffers;
- locations where vulnerable species require additional monitoring.

### Monitoring

Monitoring is appropriate when the issue is uncertain, minor, stable, or likely
to recover naturally.

### Environmental Improvement

Examples include:

- appropriate watering;
- improved drainage;
- reduced compaction;
- root protection;
- corrected mulch depth;
- reduced disturbance.

### Sanitation and Quarantine

Examples include:

- removal of infected debris;
- cleaning tools;
- appropriate disposal;
- movement restrictions;
- preventing transport of contaminated soil, wood, or nursery stock;
- creating breaks in infected host material where justified.

### Pruning

Pruning may remove dead, unsafe, damaged, or heavily infected tissue.

Timing and technique should match the host and suspected agent.

### Biological and Ecological Management

Possible approaches include:

- natural enemies;
- microbial competition;
- resistant plant material;
- habitat management;
- species diversity;
- community designs that reduce continuous host pathways.

### Chemical Treatment

Fungicides, bactericides, insecticides, or other regulated products should not
be recommended from uncertain visual identification alone.

### Prescribed Fire

Prescribed burning can be an ecological management tool in ecosystems adapted
to fire.

Depending on the ecosystem, target organism, and timing, it may:

- remove infected or susceptible surface vegetation;
- consume some infected litter or debris;
- reduce continuous plant material through which an agent may spread;
- create temporary spatial breaks or buffers;
- alter competing vegetation;
- recycle nutrients;
- restore fire-dependent ecological processes.

Prescribed fire does not sterilize an ecosystem and should not be assumed to
eliminate every pathogen.

Some organisms survive in:

- soil;
- roots;
- protected wood;
- living hosts;
- nearby untreated areas.

Fire may also injure desired plants, spread smoke, alter habitat, or create
conditions favourable to a different disturbance.

It is therefore not a general treatment for urban tree disease.

It requires:

- legal authority;
- trained professionals;
- a fire-adapted site;
- a clearly defined ecological objective;
- site-specific planning;
- suitable weather;
- containment and safety plans;
- evidence that expected benefits exceed the risks.

### Removal

Removal may be considered when:

- structural risk is unacceptable;
- decline is irreversible;
- transmission risk is high;
- treatment is impractical;
- professional guidance recommends it.

Removal may involve:

- one hazardous tree;
- infected branches;
- infected debris;
- selected host plants forming a transmission bridge;
- a carefully defined containment area.

Management should be proportional.

The goal is usually prevention, suppression, containment, resilience, or
protection of vulnerable hosts rather than elimination of every organism
associated with disease.


## 34. Evaluating Treatment Effectiveness

A treatment should be treated as an intervention that can be evaluated.

For a treated group and comparison group:

$$
\text{Treatment Effect}
=
\Delta S_{\text{treated}}
-
\Delta S_{\text{comparison}}
$$

Where:

- $\Delta S_{\text{treated}}$ = change in severity among treated plants;
- $\Delta S_{\text{comparison}}$ = change in severity among untreated or differently managed plants.

Useful outcomes may include:

- reduced symptom expansion;
- reduced new infections;
- improved canopy growth;
- reduced mortality;
- improved structural stability;
- reduced pest abundance;
- improved soil or moisture conditions.

Without comparison data, improvement after treatment does not necessarily prove that the treatment caused the improvement.

The plant may have recovered naturally or environmental conditions may have changed.

## 35. Ethical and Safety Limits

The investigation should avoid causing additional injury or spreading possible pathogens.

Fieldwork should not involve:

- cutting living tissue without permission;
- drilling into trees;
- removing bark;
- digging around roots;
- breaking branches;
- transferring material between plants;
- entering restricted property;
- applying treatments without authorization;
- collecting protected organisms;
- disturbing active wildlife habitat.

Basic precautions may include:

- photographing before touching;
- using clean equipment;
- avoiding contact between trees;
- cleaning reusable tools;
- recording observations non-destructively;
- reporting dangerous branches rather than approaching them.

Public trees should not be experimentally treated, pruned, sampled destructively, or altered without municipal permission.

## 36. Limits of a Low-Budget Field Study

Some variables cannot be measured directly with limited resources.

Potential limitations include:

- hidden root decay;
- internal trunk decay;
- vascular obstruction;
- microscopic organisms;
- exact fungal species;
- bacterial species;
- viral infection;
- soil pathogens;
- chemical contamination;
- historical injury;
- private maintenance records;
- complete root loss;
- tree genetics;
- underground root connections.

For example, directly measuring the percentage of infected root tissue might require excavation or destructive sampling.

That could seriously damage or kill the tree.

Instead, the study may rely on indirect indicators such as:

- crown decline;
- fungal structures near the base;
- root-collar abnormalities;
- soil conditions;
- leaning;
- reduced growth;
- professional inspection.

The notebook should distinguish between:

- directly observed variables;
- estimated variables;
- inferred variables;
- inaccessible variables.

## 37. False Positives and False Negatives

Field diagnosis can produce two major forms of error.

### False Positive

A healthy or low-concern condition is incorrectly classified as serious disease.

Examples might include:

- harmless lichen identified as a pathogen;
- normal bark shedding identified as tissue destruction;
- cosmetic leaf spots identified as whole-tree decline.

### False Negative

A serious condition is incorrectly classified as harmless.

Examples might include:

- internal decay overlooked because the crown remains green;
- root damage missed because roots are underground;
- early vascular disease missed before severe symptoms appear.

These errors can be represented as:

| Actual Condition | Classified Healthy | Classified Unhealthy |
|---|---|---|
| Healthy or Low Concern | Correct Negative | False Positive |
| Unhealthy or Serious | False Negative | Correct Positive |

The consequences of each error may differ.

A false positive may create unnecessary concern or treatment.

A false negative may allow a dangerous or spreading condition to remain undetected.

## 38. Bias in Observation

Several forms of bias may influence the study.

### Selection Bias

Damaged trees are more likely to be noticed and recorded.

### Confirmation Bias

Evidence supporting an expected explanation receives more attention than contradictory evidence.

### Availability Bias

A recently learned disease may be assigned too frequently because it is easy to remember.

### Observer Bias

Different observers may score the same tree differently.

### Detection Bias

Large, colourful, or dramatic symptoms are easier to detect than hidden or subtle conditions.

### Spatial Bias

Convenient streets and parks may be sampled more often than other locations.

### Seasonal Bias

Some diseases are visible only during particular times of year.

Bias should be reduced through:

- predefined sampling;
- standardized field cards;
- repeated observations;
- photographs;
- multiple observers;
- written case definitions;
- blind scoring where possible;
- preservation of uncertain classifications.

## 39. Distinguishing Correlation from Cause

Trees near a road may show greater damage than trees in a park.

This pattern could be associated with:

- road salt;
- soil compaction;
- heat;
- air pollution;
- restricted roots;
- pruning;
- vehicle damage;
- species selection;
- tree age;
- irrigation differences.

A statistical association does not by itself identify which factor caused the condition.

In general:

$$
P(Y \mid X)
\neq
P(Y \mid \text{intervention on } X)
$$

Observing that disease is common where a factor occurs is not the same as proving that changing the factor would change disease.

Causal claims require stronger evidence, such as:

- temporal order;
- biological mechanism;
- comparison groups;
- controlled intervention;
- repeated patterns;
- exclusion of major alternative explanations.

## 40. Climate Change and Emerging Conditions

Changing climate conditions may influence urban plant health through:

- warmer average temperatures;
- extreme heat;
- drought;
- intense rainfall;
- flooding;
- milder winters;
- altered insect survival;
- altered fungal development;
- longer growing seasons;
- changing host ranges;
- changing vector ranges;
- new interactions between organisms.

Climate change may increase the probability of some diseases or pests occurring in new locations.

However, an unusual local observation should not automatically be attributed to climate change.

The study should ask:

- Is the organism confirmed?
- Was it previously absent from the region?
- Is its range changing?
- Are similar changes reported elsewhere?
- Are long-term records available?
- Could ordinary local conditions explain the observation?

Climate change should be treated as a possible background driver that alters probabilities, not as an automatic diagnosis for every new symptom.

## 41. From Suspicion to Defensible Conclusion

A defensible conclusion should summarize:

1. what was observed;
2. how it was measured;
3. which explanations were considered;
4. which evidence supported each explanation;
5. which evidence contradicted each explanation;
6. what information remains unavailable;
7. the confidence level;
8. the seriousness of the condition;
9. whether further investigation is needed;
10. whether treatment or reporting is appropriate.

A conclusion might take the form:

**Observation:**  
Several maple leaves contained raised black lesions.

**Provisional interpretation:**  
The lesions are consistent with a common foliar fungal condition.

**Alternative explanations:**  
Other leaf-spot conditions remain possible.

**Current seriousness:**  
Leaf symptoms are visually widespread, but whole-tree decline has not been established.

**Evidence required:**  
Host confirmation, closer examination of lesion structure, seasonal monitoring, and comparison with unaffected trees.

**Confidence:**  
Moderate for the symptom category, low for the exact species identification.

This format prevents the notebook from claiming more than the evidence demonstrates.

## 43. Statistical Plant-Health Application Concept

A possible long-term result of this methodology is an interactive application
for exploring plant-health observations and statistical predictions.

The concept is independent of platform. It could eventually be implemented as
a web application, desktop program, mobile application, or another interface.
This notebook defines the scientific and mathematical behaviour beneath the
interface rather than selecting a technology stack.

The application would combine:

- an inventory of real plants;
- field observations;
- disease incidence and severity;
- environmental measurements;
- species and demographic information;
- statistical models calibrated from observed data.

It would not attempt to simulate every physical and biological process inside a
tree. Its predictions would describe patterns learned from existing data.

### Three Main Types of Controls

#### 1. Study or Outcome Selection

The user chooses what is being analyzed, such as:

- incidence;
- severity;
- susceptibility;
- progression rate;
- mortality risk;
- growth response.

Different outcomes may require different equations, datasets, and case
definitions.

#### 2. Population and Data Filters

Filters determine which observations are included.

Possible filters include:

- geographic area;
- species;
- age or size demographic;
- street, park, or neighbourhood;
- planting environment;
- symptom or disease category;
- observation period;
- soil, moisture, or exposure class.

Filtering changes the study population:

$$
\mathcal{D}
\rightarrow
\mathcal{D}_{\mathrm{maple}}
\rightarrow
\mathcal{D}_{\mathrm{maple,young}}
\rightarrow
\mathcal{D}_{\mathrm{maple,young,park\ A}}
$$

#### 3. Variable Selection and Sliders

Variable selectors determine which predictors remain explicit in the model.
Sliders adjust the values supplied to those predictors.

Possible slider variables include:

- temperature;
- rainfall;
- soil moisture;
- drought index;
- tree age;
- planting density;
- distance from road;
- wound severity;
- canopy loss.

For a fitted model:

$$
\widehat{Y}_i
=
\widehat{\beta}_0
+
\sum_{j=1}^{p}
\widehat{\beta}_j x_{i,j}
$$

Where:

- $\widehat{Y}_i$ = predicted outcome for tree $i$;
- $\widehat{\beta}_j$ = coefficient estimated from observed data;
- $x_{i,j}$ = recorded or user-adjusted predictor value.

Moving a slider changes an input value. Selecting or removing a variable
changes which terms appear. Applying a filter changes which observations form
the selected population or the data used for a particular view.

### Colour-Coded Map

Each mapped tree could be assigned a colour representing the selected output,
such as lower or higher predicted susceptibility.

As a slider moves, the predicted value can be recalculated and the tree colour
can change. Filters can restrict the display to a species, demographic group,
time period, or local area.

The map could reveal:

- high-risk clusters;
- vulnerable demographic groups;
- differences among species;
- possible resistant or buffer populations;
- locations where host continuity may facilitate spread;
- places where environmental conditions correspond with higher severity.

### Statistical, Not Mechanistic, Prediction

The application would estimate conditional patterns such as:

$$
P(
\text{illness}
\mid
\text{species, age, location, environment, observations}
)
$$

It would not claim that a slider proves a physical causal mechanism.
Predictions would be limited by sample quality, representativeness,
identification accuracy, uncertainty, and the range of conditions present in
the training data.

Calibrating the coefficients resembles training a predictive model: recorded
examples are used to estimate the weights that best describe or predict the
observed outcomes. This may use conventional statistics or machine learning;
it does not require an artificial-intelligence label.

The application is therefore a decision-support and visualization system, not
a replacement for field diagnosis or expert assessment.

## 45. Core Methodological Principles

This project will follow several core principles.

1. A symptom is not automatically a diagnosis.

2. Host species, age, and seasonal stage should be identified before assigning
   a named disease.

3. Visible damage does not always indicate serious decline.

4. Serious conditions are not always visually dramatic.

5. Several causes may act together or in sequence.

6. Fungi are natural components of plant ecosystems and are not automatically
   evidence of disease.

7. The visible organism may be a primary pathogen, secondary pathogen,
   opportunist, decomposer, mutualist, or harmless associate.

8. Seasonal growth, dormancy, leaf loss, and tissue replacement must be
   distinguished from illness.

9. Roots are central to plant health, but not every disease begins in the roots.

10. Cultivation and transplanting may create stress without proving that a tree
    is permanently weak.

11. Healthy-looking trees must be sampled alongside damaged trees.

12. Opportunistic observations must be separated from statistical samples.

13. Individual, population, and community health answer different questions.

14. Species should remain an explicit variable when susceptibility may differ.

15. Incidence, severity, and biological impact are different measurements.

16. Leaves, branches, trunks, crowns, and roots should be evaluated separately.

17. Repeated observations are needed to measure progression.

18. Spatial clusters do not automatically prove transmission.

19. Correlation does not automatically establish cause.

20. Prevention and treatment should follow evidence-based diagnosis.

21. Uncertainty must be recorded rather than hidden.

22. Destructive investigation should not be performed without permission and
    appropriate expertise.

23. Strong conclusions require evidence capable of distinguishing plausible
    alternatives.

## 46. Community Ecology as a Plant-Health Framework

Population ecology examines individuals of the same species.

A municipal inventory can be treated operationally as one large study
population and divided into demographic subpopulations. When several species
are included, those species populations together form an ecological community.
Community ecology examines how their traits, distributions, abundance, and
interactions affect the larger system.

An urban tree community may be represented as:

$$
\mathcal{C}
=
\{
P_1,
P_2,
\dots,
P_S
\}
$$

Where:

- $\mathcal{C}$ = the local plant community;
- $P_s$ = population of species $s$;
- $S$ = number of species included.

The total number of trees is:

$$
N
=
\sum_{s=1}^{S}
N_s
$$

But the total alone is not sufficient.

Two communities can contain the same number of trees while having very
different:

- species composition;
- susceptibility;
- age structure;
- spacing;
- disease incidence;
- transmission pathways;
- resilience.

Community analysis therefore retains the species index:

$$
Y_{i,s}
$$

instead of immediately reducing every observation to:

$$
Y_i
$$

### Comparing Species Under Similar Conditions

Suppose two species occupy the same local area and experience similar weather.

If one species has a higher incidence or faster progression rate, that pattern
may indicate:

- greater susceptibility;
- lower tolerance;
- different defence mechanisms;
- a more favourable relationship with a vector;
- different symptom visibility;
- different planting density;
- a hidden environmental difference.

The comparison becomes stronger when age, location, and exposure are similar.

For example:

$$
\Delta g
=
g_{s_1}
-
g_{s_2}
$$

Where:

- $g_{s_1}$ = progression rate in species $s_1$;
- $g_{s_2}$ = progression rate in species $s_2$.

A large value of $\Delta g$ identifies a difference that requires
explanation.

It does not by itself prove the mechanism.

### Community Roles in Spread

The arrangement of species may affect transmission.

A susceptible species planted continuously may create a connected host path.

A resistant or non-host species planted between susceptible hosts may reduce
direct continuity.

A tolerant host may either support resilience or conceal transmission,
depending on whether it carries the agent.

The community therefore contains possible:

- source populations;
- sink populations;
- reservoir populations;
- bridge hosts;
- highly susceptible hosts;
- tolerant hosts;
- non-hosts;
- buffer species.

These roles can guide:

- conservation priorities;
- surveillance;
- planting diversity;
- spacing;
- replacement decisions;
- containment strategies;
- protection of rare or vulnerable species.

The ecological goal is not simply to minimize every visible interaction.

It is to maintain a resilient community while preventing particular agents or
conditions from causing unacceptable loss.

## 47. Selecting, Aggregating, and Reintroducing Variables

A useful model should be detailed enough to answer the question but simple
enough to estimate and interpret.

Let the complete set of possible predictors be:

$$
\mathbf{X}_i
=
[
X_{i,1},
X_{i,2},
\dots,
X_{i,p}
]
$$

Possible predictors include:

- species;
- age;
- temperature;
- rainfall;
- soil moisture;
- location;
- planting density;
- road exposure;
- wounds;
- management;
- neighbouring species;
- season.

A model does not need to use every available variable at once.

Define a selection indicator:

$$
z_j
=
\begin{cases}
1, & \text{if variable } j \text{ is included} \\
0, & \text{if variable } j \text{ is excluded}
\end{cases}
$$

Then:

$$
\widehat{Y}_i
=
\widehat{\beta}_0
+
\sum_{j=1}^{p}
z_j
\widehat{\beta}_j
X_{i,j}
$$

The value of $z_j$ represents whether a variable is brought to the front of
the current analysis.

This gives a mathematical interpretation to adding or removing model
variables.

### Filtering Is Different From Variable Selection

Filtering changes which rows are used:

$$
\mathcal{D}^{\ast}
=
\{
i \in \mathcal{D}
:
F(i)=1
\}
$$

Where:

- $\mathcal{D}$ = complete dataset;
- $F(i)$ = filter rule;
- $\mathcal{D}^{\ast}$ = selected observations.

Variable selection changes which columns or predictors are used:

$$
\mathbf{X}^{\ast}
\subseteq
\mathbf{X}
$$

A study can therefore:

1. filter to maple trees in one park;
2. include age, moisture, and temperature as predictors;
3. hold species and location constant;
4. estimate how the remaining variables relate to susceptibility.

### Aggregating Variables

Several minor variables may be combined:

$$
B_i
=
\sum_{r=1}^{R}
\gamma_r U_{i,r}
$$

Where:

- $U_{i,r}$ = variables not interpreted separately;
- $\gamma_r$ = their combined weights;
- $B_i$ = aggregated background contribution.

The main model becomes:

$$
\widehat{Y}_i
=
\widehat{\beta}_0
+
\sum_{j=1}^{p}
z_j
\widehat{\beta}_j
X_{i,j}
+
B_i
$$

If one background variable later proves important, it can be removed from
$B_i$ and added explicitly to the main sum.

### Adding Group Indices

When the question changes from individuals to demographic subpopulations,
additional indices can be added:

$$
Y_{i,s,a,l,t}
$$

Where:

- $i$ = individual;
- $s$ = species;
- $a$ = age class;
- $l$ = location class;
- $t$ = time.

A group mean can be written as:

$$
\bar{Y}_{s,a,l,t}
=
\frac{1}{n_{s,a,l,t}}
\sum_{i=1}^{n_{s,a,l,t}}
Y_{i,s,a,l,t}
$$

This allows the same dataset to support many studies without requiring every
study to collect every possible variable again.

The value of detailed data is therefore not that every variable must always be
used.

Its value is that later studies can choose which variables to:

- filter;
- compare;
- hold constant;
- aggregate;
- model;
- display.

## 48. Application Controls as Mathematical Operations

The proposed application, regardless of whether it is eventually a web,
mobile, desktop, or other application, can be understood as an interface for
manipulating data selection and statistical equations.

The controls belong to three mathematical layers.

### Layer 1: Select the Study

The user chooses the outcome:

$$
Y
\in
\{
\text{incidence},
\text{severity},
\text{susceptibility},
\text{progression},
\text{mortality},
\text{growth}
\}
$$

The selected outcome determines which model is used.

### Layer 2: Filter the Population

The user defines the group:

$$
\mathcal{D}^{\ast}
=
\mathcal{D}
\cap
\mathcal{S}
\cap
\mathcal{A}
\cap
\mathcal{L}
\cap
\mathcal{T}
$$

Where:

- $\mathcal{S}$ = selected species;
- $\mathcal{A}$ = selected age classes;
- $\mathcal{L}$ = selected locations;
- $\mathcal{T}$ = selected time period.

These controls determine which trees are being studied.

### Layer 3: Select and Adjust Predictors

Variable controls determine which terms appear:

$$
\widehat{Y}_i
=
\widehat{\beta}_0
+
\sum_{j=1}^{p}
z_j
\widehat{\beta}_j
x_{i,j}
$$

Sliders change selected values of $x_{i,j}$.

For example:

- temperature slider changes $x_{\mathrm{temperature}}$;
- rainfall slider changes $x_{\mathrm{rainfall}}$;
- age slider changes $x_{\mathrm{age}}$;
- density slider changes $x_{\mathrm{density}}$.

Filters change the dataset $\mathcal{D}^{\ast}$.

The controls therefore perform three different operations:

$$
\boxed{
\text{Choose Outcome}
\rightarrow
\text{Filter Population}
\rightarrow
\text{Select Variables}
\rightarrow
\text{Adjust Values}
}
$$

### Colour-Coded Map Output

For every mapped tree:

$$
\widehat{p}_i
=
P(
Y_i=1
\mid
\mathbf{x}_i,
\mathcal{D}^{\ast}
)
$$

A colour scale can represent $\widehat{p}_i$.

As a slider moves, the model recalculates $\widehat{p}_i$, and the colour of
each tree changes.

The map may therefore show trees becoming statistically more or less
susceptible under different selected conditions.

This is not a physics simulation.

It is a statistical decision-support system based on:

- observed plants;
- recorded disease outcomes;
- measured environmental conditions;
- estimated relationships;
- clearly stated uncertainty.

The application is the practical endpoint of the methodology developed in
this notebook.

The notebook explains what the app must calculate beneath the interface and
what evidence is required before its outputs can be trusted.

- host identification;
- symptom and sign description;
- seasonal and environmental context;
- biological mechanisms;
- competing hypotheses;
- individual, demographic, population, and community sampling;
- cross-species comparisons;
- severity and incidence measurements;

$$
\boxed{
\text{Observe}
\rightarrow
\text{Identify the Host}
\rightarrow
\text{Describe}
\rightarrow
\text{Classify}
\rightarrow
\text{Measure}
}
$$

$$
\boxed{
\rightarrow
\text{Compare Hypotheses}
\rightarrow
\text{Collect Evidence}
\rightarrow
\text{Evaluate Seriousness}
}
$$

$$
\boxed{
\rightarrow
\text{Compare Populations and Species}
\rightarrow
\text{Conclude}

}
$$

This framework allows unusual possibilities to be investigated without being

for severe disease.

The purpose is not merely to create a list of plant diseases.

It is to create a systematic framework for deciding:

- what happened;
- which causes are plausible;
- how confidently the condition can be identified;
- how serious it is for one plant;
- how common it is within one species or demographic subpopulation;
- how incidence and progression differ among species;
- how community composition influences resilience or spread;
- whether a species acts as a susceptible host, carrier, bridge, or buffer;
- whether the condition is changing;

- dataset selector;
- study selector;
- filter;
- variable control;
- slider;
- map colour;
- predicted susceptibility value;

It would not rely on unsupported guesswork or pretend to reproduce every
physical process inside a plant.

The notebook is therefore:

- a field methodology;

- a statistical modelling guide;
- an early technical specification for a future plant-health application.
